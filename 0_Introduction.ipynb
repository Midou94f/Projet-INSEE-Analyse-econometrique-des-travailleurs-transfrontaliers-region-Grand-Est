## 1. Le Grand Est : première région frontalière de France

Avec ses 450 kilomètres de frontières partagées avec quatre pays — le Luxembourg, l'Allemagne, la Belgique et la Suisse —, la région Grand Est constitue le principal territoire français concerné par les mobilités professionnelles transfrontalières. En 2021, selon le Recensement de la Population de l'INSEE, plus de **203 000 habitants** de la région traversent quotidiennement une frontière pour exercer leur activité professionnelle, faisant du travail frontalier un phénomène économique et social de première importance.

La répartition géographique de ces flux révèle une forte polarisation territoriale. Le **Luxembourg** capte à lui seul la moitié des travailleurs frontaliers du Grand Est, suivi par l'**Allemagne** (près d'un quart), la **Suisse** (plus de 20 %) et la **Belgique** (4 %). Cette concentration s'explique par les différentiels de salaires considérables entre la France et ses voisins — particulièrement marqués avec le Luxembourg et la Suisse — ainsi que par le dynamisme des marchés du travail de ces pays. À titre d'illustration, le taux de chômage suisse oscille autour de 4 %, bien en deçà des 7,3 % enregistrés dans le Grand Est en 2024.

À l'échelle infrarégionale, trois départements concentrent l'essentiel des flux sortants :

| Département | Destination principale | Part des frontaliers dans l'emploi local |
|-------------|------------------------|:----------------------------------------:|
| **Moselle** | Luxembourg | Jusqu'à 50 % dans la zone de Thionville |
| **Haut-Rhin** | Suisse, Allemagne | Jusqu'à 50 % dans la zone de Saint-Louis |
| **Meurthe-et-Moselle** | Luxembourg | Concentration sur le nord du département |

Dans les zones d'emploi de Thionville et de Saint-Louis, la moitié des actifs occupés sont des travailleurs frontaliers — une proportion sans équivalent sur le territoire national. Cette intensité du phénomène frontalier façonne profondément l'économie locale, les marchés immobiliers, les infrastructures de transport et l'organisation sociale de ces territoires.

## 2. Un phénomène en croissance soutenue

Le travail frontalier dans le Grand Est n'est pas un phénomène statique : il s'inscrit dans une dynamique de croissance continue depuis plus de deux décennies. Entre 1999 et 2021, le nombre de frontaliers vers le Luxembourg a **doublé**, tandis que les flux vers la Suisse et la Belgique ont également connu une progression marquée. Seule l'Allemagne fait exception, avec une diminution des effectifs frontaliers, notamment dans les professions peu qualifiées.

Cette évolution s'accompagne d'une transformation qualitative des profils. Les travailleurs frontaliers sont aujourd'hui **plus diplômés** qu'il y a vingt ans : le nombre de ceux exerçant une profession hautement qualifiée a été multiplié par cinq vers le Luxembourg et par 3,4 vers la Suisse. Les cadres des services administratifs, les ingénieurs informatiques et les personnels de recherche représentent désormais une part croissante des navetteurs quotidiens.

Cette attractivité persistante des marchés du travail voisins soulève des **enjeux majeurs** pour les territoires français concernés :

- **Enjeux économiques** : tensions sur les marchés du travail locaux, difficultés de recrutement pour les entreprises françaises, transferts de revenus vers les territoires frontaliers
- **Enjeux fiscaux** : imposition des revenus à l'étranger, conventions bilatérales, compensation aux collectivités locales
- **Enjeux de transport** : congestion des axes routiers (A31, A35), saturation des parkings relais, émissions de CO₂ (les trajets d'un frontalier génèrent en moyenne deux fois plus d'émissions que ceux d'un non-frontalier)
- **Enjeux sociaux et familiaux** : conciliation vie professionnelle et vie personnelle, garde d'enfants adaptée aux horaires atypiques, organisation conjugale des mobilités

Les frontaliers du Grand Est parcourent en moyenne **38 kilomètres** pour se rendre au travail, contre 15 kilomètres pour les travailleurs non frontaliers — soit une distance 2,5 fois supérieure. Vers le Luxembourg, cette distance moyenne atteint 42 kilomètres, et un quart des navetteurs effectuent plus de 100 kilomètres aller-retour quotidiennement. La voiture reste le mode de transport privilégié (96 % des frontaliers vers l'Allemagne et la Belgique), le train ne captant qu'une part marginale des déplacements (12 % vers le Luxembourg).

## 3. Un phénomène économiquement significatif mais insuffisamment compris

Malgré son ampleur — le Grand Est compte plus de frontaliers que certaines régions ne comptent de chômeurs —, le travail transfrontalier demeure un objet d'étude relativement peu exploré dans la littérature économétrique française. Si les statistiques descriptives produites par l'INSEE documentent avec précision les flux, les distances et les secteurs d'activité, les **déterminants individuels** de la décision de travail frontalier restent largement méconnus.

Pourquoi, à caractéristiques observables comparables, certains individus choisissent-ils de travailler à l'étranger tandis que d'autres restent sur le marché du travail français ? Quels facteurs — au-delà de la simple proximité géographique — influencent cette probabilité ? Le genre, le diplôme, la situation familiale, le secteur d'activité jouent-ils un rôle, et si oui, dans quelle mesure ?

Plus fondamentalement encore, la **dimension familiale** du travail frontalier n'a, à notre connaissance, jamais fait l'objet d'une analyse économétrique sur données françaises. Or, la décision de traverser quotidiennement une frontière pour travailler ne peut être appréhendée comme un choix purement individuel : elle s'inscrit dans un contexte conjugal et familial qui conditionne les arbitrages en termes de temps, de revenus et d'organisation domestique. Les choix professionnels des deux membres d'un couple sont-ils coordonnés ? Observe-t-on des stratégies de spécialisation au sein des ménages, où l'un des conjoints privilégie l'emploi frontalier tandis que l'autre assure une présence domestique plus importante ?

C'est précisément à ces questions que le présent travail de recherche entend apporter des éléments de réponse.

## 4. Problématique et questions de recherche

### 4.1 Problématique générale

Ce travail de recherche s'inscrit dans une démarche d'analyse quantitative des comportements de mobilité professionnelle transfrontalière. Il vise à répondre à la problématique suivante :

> **Quels sont les déterminants individuels et familiaux de la décision de travail transfrontalier dans la région Grand Est, et dans quelle mesure ces déterminants varient-ils selon le genre et s'inscrivent-ils dans une logique de coordination conjugale ?**

Cette problématique se décline en trois questions de recherche complémentaires, chacune correspondant à un niveau d'analyse distinct et à une modélisation économétrique spécifique.

---

### 4.2 Trois questions de recherche articulées

#### Question 1 — Les déterminants individuels

*Quels facteurs individuels influencent la probabilité d'être travailleur transfrontalier ?*

Cette première question vise à identifier et quantifier l'effet des caractéristiques sociodémographiques, professionnelles et géographiques sur la probabilité d'exercer une activité professionnelle hors de France. Elle constitue le socle de l'analyse, en établissant les **effets moyens** des différents déterminants sur l'ensemble de la population active du Grand Est.

Les dimensions explorées incluent :
- La **géographie** : effet du département de résidence, proxy de la distance à la frontière
- Le **capital humain** : niveau de diplôme, catégorie socioprofessionnelle
- Les **caractéristiques professionnelles** : secteur d'activité, type de contrat, temps de travail
- La **situation familiale** : présence d'enfants, statut conjugal
- L'**origine** : nationalité, lieu de naissance
- Les **ressources du ménage** : motorisation, type de logement

---

#### Question 2 — L'hétérogénéité selon le genre

*Les déterminants du travail transfrontalier agissent-ils différemment pour les hommes et pour les femmes ?*

Cette deuxième question interroge l'existence d'une **hétérogénéité des effets selon le genre**. L'hypothèse sous-jacente est que les contraintes et les arbitrages auxquels font face les hommes et les femmes diffèrent, notamment en matière de conciliation entre vie professionnelle et responsabilités familiales.

Les mécanismes testés incluent :
- L'**effet différencié de la parentalité** : la présence d'enfants pénalise-t-elle davantage les femmes que les hommes dans l'accès au travail frontalier ?
- Le **rôle du temps partiel** : le temps partiel constitue-t-il une stratégie de conciliation pour les femmes, ou une barrière à l'emploi frontalier ?
- La **segmentation sectorielle** : les effets du secteur d'activité sont-ils symétriques selon le genre ?
- L'**accès aux ressources du ménage** : la motorisation joue-t-elle un rôle différent pour les hommes et les femmes ?

---

#### Question 3 — La coordination au sein des couples

*Les décisions de travail transfrontalier des deux membres d'un couple sont-elles coordonnées ou indépendantes ?*

Cette troisième question constitue l'apport le plus original de ce travail. Elle dépasse l'analyse individuelle pour appréhender le travail frontalier comme une **décision de ménage**. L'enjeu est de déterminer si les choix professionnels des conjoints sont indépendants ou s'ils s'inscrivent dans une logique de coordination — explicite ou implicite — pouvant prendre la forme d'une complémentarité (les deux conjoints sont frontaliers ensemble) ou d'une substitution (un seul des deux est frontalier, l'autre assurant une présence domestique).

Les mécanismes explorés incluent :
- Les **effets de réseau informationnels** : un conjoint déjà frontalier facilite-t-il l'accès de l'autre au marché du travail étranger ?
- Les **préférences communes non observées** : les couples partagent-ils des caractéristiques inobservables (tolérance aux trajets longs, ouverture internationale) qui génèrent une corrélation des choix ?
- La **coordination logistique** : covoiturage, synchronisation des horaires, mutualisation des contraintes de garde d'enfants
- La **stratégie de maximisation du revenu du ménage** : les deux conjoints arbitrent-ils conjointement pour optimiser le revenu familial total ?

---

### 4.3 Posture épistémologique : une démarche positiviste

Dans ce travail, nous adoptons une posture strictement **positiviste**. Plutôt que de formuler des hypothèses a priori que nous chercherions à confirmer ou infirmer — démarche hypothético-déductive classique —, nous choisissons de **naviguer au gré des données et des résultats empiriques**, en laissant les faits statistiques guider notre compréhension du phénomène.

Cette approche inductive présente un avantage majeur : elle évite le biais de confirmation qui consisterait à orienter l'analyse vers la validation d'un postulat préétabli. Les modèles économétriques sont ici mobilisés comme des **instruments d'exploration** plutôt que comme des outils de test d'hypothèses prédéfinies. Nous laissons les coefficients estimés, les effets marginaux et les paramètres structurels (notamment le paramètre de corrélation ρ du probit bivarié) révéler les régularités empiriques, sans présumer de leur signe ou de leur magnitude.

Néanmoins, la littérature existante et les analyses descriptives disponibles suggèrent certaines tendances que nos résultats pourront confirmer, nuancer ou contredire :

| Ce que suggère la littérature | Question associée |
|-------------------------------|:-----------------:|
| La proximité géographique à la frontière serait le déterminant principal de la probabilité de travail transfrontalier, devant les caractéristiques individuelles | Q1 |
| Les contraintes de conciliation famille-travail pèseraient différemment sur les hommes et les femmes, notamment via la parentalité et le temps partiel | Q2 |
| Les décisions professionnelles au sein des couples ne seraient pas indépendantes : on observerait une surreprésentation des couples « bi-frontaliers » par rapport à l'hypothèse d'indépendance statistique | Q3 |

Ces intuitions issues de la littérature ne constituent en aucun cas des hypothèses que nous cherchons à valider. Elles servent uniquement de points de repère pour situer nos résultats dans le paysage des connaissances existantes. Notre démarche reste résolument **empirique et ouverte** : si les données contredisent ces intuitions, nous en rendrons compte avec la même rigueur que si elles les confirmaient.

## 5. Démarche économétrique : une famille de modèles probit

### 5.1 Justification du choix méthodologique

Pour répondre aux trois questions de recherche, nous mobilisons une **famille cohérente de modèles économétriques** : les modèles probit. Ce choix, loin d'être arbitraire, répond à trois exigences méthodologiques fondamentales.

**Premièrement, la cohérence d'interprétation.** Les modèles probit et logit, bien que souvent présentés comme équivalents, reposent sur des distributions sous-jacentes différentes (normale vs logistique). Leurs coefficients ne sont pas directement comparables. En utilisant exclusivement des modèles probit tout au long de l'analyse, nous préservons une **même échelle d'estimation** et une **même logique d'interprétation**, autorisant des comparaisons rigoureuses entre les résultats des différentes parties.

**Deuxièmement, la compatibilité avec le modèle structurel final.** La troisième question de recherche — relative à la coordination des décisions au sein des couples — requiert l'estimation d'un **probit bivarié**, seul outil économétrique permettant de modéliser simultanément deux variables binaires corrélées. Il n'existe pas d'équivalent opérationnel en spécification logistique (le « logit bivarié » n'est pas identifié sans hypothèses supplémentaires fortes). Dès lors, pour maintenir la cohérence de l'ensemble du dispositif analytique, il est méthodologiquement préférable d'adopter le probit dès les premières étapes.

**Troisièmement, la lisibilité des effets marginaux.** Les modèles probit permettent le calcul d'effets marginaux moyens (AME) directement interprétables en **points de pourcentage**, facilitant la communication des résultats et leur traduction en termes de politique publique.

---

### 5.2 Modèle 1 — Probit binaire simple (Partie 3)

Le premier modèle estime la probabilité individuelle d'exercer une activité professionnelle hors de France. Il s'agit d'un modèle probit binaire classique.

#### Variable dépendante

Soit $Y_i$ la variable binaire indiquant le statut de travailleur transfrontalier :

$$Y_i = \begin{cases} 1 & \text{si l'individu } i \text{ travaille à l'étranger (LU, DE, CH, BE)} \\ 0 & \text{sinon} \end{cases}$$

#### Spécification

Le modèle repose sur une variable latente $Y_i^*$ représentant la propension inobservée au travail transfrontalier :

$$Y_i^* = X_i'\beta + \varepsilon_i \quad \text{avec} \quad \varepsilon_i \sim \mathcal{N}(0, 1)$$

où :
- $X_i$ est le vecteur des caractéristiques individuelles et du ménage (dimension $K \times 1$)
- $\beta$ est le vecteur des coefficients à estimer (dimension $K \times 1$)
- $\varepsilon_i$ est le terme d'erreur, supposé suivre une loi normale centrée réduite

La règle d'observation est :

$$Y_i = \mathbb{1}_{[Y_i^* > 0]}$$

#### Probabilité et vraisemblance

La probabilité d'être transfrontalier s'écrit :

$$P(Y_i = 1 \mid X_i) = P(Y_i^* > 0) = P(\varepsilon_i > -X_i'\beta) = \Phi(X_i'\beta)$$

où $\Phi(\cdot)$ désigne la fonction de répartition de la loi normale centrée réduite.

La log-vraisemblance du modèle est :

$$\ln \mathcal{L}(\beta) = \sum_{i=1}^{n} \left[ Y_i \ln \Phi(X_i'\beta) + (1 - Y_i) \ln (1 - \Phi(X_i'\beta)) \right]$$

Les coefficients $\hat{\beta}$ sont estimés par **maximum de vraisemblance** (MLE).

#### Effets marginaux

Les coefficients $\beta$ ne mesurent pas directement l'effet d'une variable sur la probabilité. L'**effet marginal moyen** (AME) d'une variable continue $X_k$ est :

$$\text{AME}_k = \frac{1}{n} \sum_{i=1}^{n} \phi(X_i'\hat{\beta}) \cdot \hat{\beta}_k$$

où $\phi(\cdot)$ est la densité de la loi normale centrée réduite.

Pour une variable binaire $D$, l'effet marginal est la différence de probabilités prédites :

$$\text{AME}_D = \frac{1}{n} \sum_{i=1}^{n} \left[ \Phi(X_i'\hat{\beta} \mid D=1) - \Phi(X_i'\hat{\beta} \mid D=0) \right]$$

### 5.3 Modèle 2 — Probit avec interactions genre (Partie 4)

Le deuxième modèle enrichit la spécification précédente en introduisant des **termes d'interaction** entre le genre et l'ensemble des variables explicatives, afin de tester l'hétérogénéité des effets.

#### Spécification

Soit $F_i$ une variable indicatrice égale à 1 si l'individu est une femme. Le modèle s'écrit :

$$Y_i^* = X_i'\beta + F_i \cdot \gamma_0 + (F_i \times X_i)'\gamma + \varepsilon_i$$

soit, de manière développée :

$$Y_i^* = \beta_0 + \sum_{k=1}^{K} \beta_k X_{ik} + \gamma_0 F_i + \sum_{k=1}^{K} \gamma_k (F_i \times X_{ik}) + \varepsilon_i$$

où :
- $\beta_k$ mesure l'effet de la variable $X_k$ pour les **hommes** (catégorie de référence)
- $\gamma_0$ mesure l'écart de niveau entre femmes et hommes (à caractéristiques égales)
- $\gamma_k$ mesure l'**écart d'effet** de la variable $X_k$ entre femmes et hommes

#### Interprétation des coefficients

Pour les **hommes** ($F_i = 0$) :
$$\frac{\partial Y_i^*}{\partial X_k} = \beta_k$$

Pour les **femmes** ($F_i = 1$) :
$$\frac{\partial Y_i^*}{\partial X_k} = \beta_k + \gamma_k$$

Le coefficient d'interaction $\gamma_k$ répond donc directement à la question : *« L'effet de $X_k$ est-il significativement différent pour les femmes ? »*

#### Test d'hétérogénéité

L'hypothèse nulle d'homogénéité des effets selon le genre s'écrit :

$$H_0 : \gamma_0 = \gamma_1 = \cdots = \gamma_K = 0$$

Elle est testée par un **test de Wald joint** ou un **test du rapport de vraisemblance** (LR) comparant le modèle avec interactions au modèle sans interactions.

### 5.4 Modèle 3 — Probit bivarié (Partie 5)

Le troisième modèle constitue l'apport méthodologique le plus substantiel de ce travail. Il modélise **simultanément** les décisions des deux membres d'un couple, en autorisant une corrélation entre les termes d'erreur des deux équations.

#### Système d'équations

Pour chaque couple $c$, on observe deux décisions binaires : celle de la personne de référence ($Y_{REF,c}$) et celle du conjoint ($Y_{CONJ,c}$). Le modèle spécifie deux équations de variables latentes :

$$\begin{cases} Y_{REF,c}^* = X_{REF,c}'\beta_{REF} + Z_c'\gamma_{REF} + \varepsilon_{REF,c} \\[6pt] Y_{CONJ,c}^* = X_{CONJ,c}'\beta_{CONJ} + Z_c'\gamma_{CONJ} + \varepsilon_{CONJ,c} \end{cases}$$

où :
- $X_{REF,c}$ et $X_{CONJ,c}$ sont les vecteurs de caractéristiques **individuelles** de chaque conjoint
- $Z_c$ est le vecteur des caractéristiques **du ménage**, communes aux deux équations (département, nombre d'enfants, motorisation, logement)
- $\beta_{REF}$, $\beta_{CONJ}$, $\gamma_{REF}$, $\gamma_{CONJ}$ sont les vecteurs de coefficients spécifiques à chaque équation

#### Distribution jointe des erreurs

L'originalité du probit bivarié réside dans la spécification de la distribution jointe des termes d'erreur :

$$\begin{pmatrix} \varepsilon_{REF} \\ \varepsilon_{CONJ} \end{pmatrix} \sim \mathcal{N}_2 \left( \begin{pmatrix} 0 \\ 0 \end{pmatrix}, \begin{pmatrix} 1 & \rho \\ \rho & 1 \end{pmatrix} \right)$$

Le paramètre $\rho \in [-1, 1]$ est le **coefficient de corrélation** entre les résidus des deux équations. C'est le paramètre central du modèle.

#### Interprétation du paramètre ρ

| Valeur de ρ | Interprétation | Mécanisme économique |
|:-----------:|----------------|----------------------|
| $\rho > 0$ | Corrélation positive | **Coordination** : les deux conjoints tendent à être transfrontaliers ensemble |
| $\rho < 0$ | Corrélation négative | **Substitution** : si l'un est transfrontalier, l'autre tend à ne pas l'être |
| $\rho = 0$ | Indépendance | **Pas de coordination** : les décisions sont prises indépendamment |

Cruciale est la nuance suivante : $\rho$ mesure la corrélation **résiduelle**, c'est-à-dire la corrélation entre les décisions **après contrôle** de toutes les caractéristiques observables (département, diplôme, secteur, etc.). Un $\rho$ significativement différent de zéro indique donc l'existence de facteurs inobservés communs ou d'une coordination explicite entre les conjoints.

#### Vraisemblance

La contribution à la vraisemblance de chaque couple dépend du quadrant observé $(Y_{REF}, Y_{CONJ}) \in \{0,1\}^2$ :

$$\mathcal{L}_c = \begin{cases} \Phi_2(\eta_{REF}, \eta_{CONJ}; \rho) & \text{si } (Y_{REF}, Y_{CONJ}) = (1, 1) \\[4pt] \Phi_2(\eta_{REF}, -\eta_{CONJ}; -\rho) & \text{si } (Y_{REF}, Y_{CONJ}) = (1, 0) \\[4pt] \Phi_2(-\eta_{REF}, \eta_{CONJ}; -\rho) & \text{si } (Y_{REF}, Y_{CONJ}) = (0, 1) \\[4pt] \Phi_2(-\eta_{REF}, -\eta_{CONJ}; \rho) & \text{si } (Y_{REF}, Y_{CONJ}) = (0, 0) \end{cases}$$

où :
- $\eta_{REF} = X_{REF}'\beta_{REF} + Z'\gamma_{REF}$ et $\eta_{CONJ} = X_{CONJ}'\beta_{CONJ} + Z'\gamma_{CONJ}$ sont les index linéaires
- $\Phi_2(\cdot, \cdot; \rho)$ est la fonction de répartition de la loi normale bivariée avec corrélation $\rho$

La log-vraisemblance totale est :

$$\ln \mathcal{L}(\beta_{REF}, \beta_{CONJ}, \gamma_{REF}, \gamma_{CONJ}, \rho) = \sum_{c=1}^{C} \ln \mathcal{L}_c$$

L'estimation est réalisée par **maximum de vraisemblance**, avec calcul numérique de la CDF normale bivariée (méthode de Drezner-Wesolowsky avec quadrature de Gauss-Legendre).

#### Test de l'hypothèse d'indépendance

L'hypothèse nulle d'indépendance des décisions est :

$$H_0 : \rho = 0$$

Elle est testée par :
- Un **test de Wald** : $z = \hat{\rho} / \text{SE}(\hat{\rho})$
- Un **test du rapport de vraisemblance** : $LR = 2 \left[ \ln \mathcal{L}(\hat{\rho}) - \ln \mathcal{L}(\rho = 0) \right] \sim \chi^2(1)$

Le rejet de $H_0$ constitue la preuve statistique d'une coordination intra-ménage.

### 5.5 Articulation des trois modèles

Les trois modèles s'articulent dans une **progression logique** :

```
┌─────────────────────────────────────────────────────────────────────────────┐
│                                                                             │
│   PROBIT BINAIRE SIMPLE (Partie 3)                                          │
│   ─────────────────────────────────                                         │
│   • Unité d'analyse : individu                                              │
│   • Question : Quels sont les déterminants moyens ?                         │
│   • Output : coefficients β, effets marginaux AME                           │
│                         │                                                   │
│                         ▼                                                   │
│   PROBIT AVEC INTERACTIONS (Partie 4)                                       │
│   ────────────────────────────────────                                      │
│   • Unité d'analyse : individu × genre                                      │
│   • Question : Les effets sont-ils hétérogènes selon le genre ?             │
│   • Output : coefficients d'interaction γ, tests de Wald                    │
│                         │                                                   │
│                         ▼                                                   │
│   PROBIT BIVARIÉ (Partie 5)                                                 │
│   ─────────────────────────                                                 │
│   • Unité d'analyse : couple                                                │
│   • Question : Les décisions des conjoints sont-elles coordonnées ?         │
│   • Output : paramètre ρ, probabilités conditionnelles                      │
│                                                                             │
└─────────────────────────────────────────────────────────────────────────────┘
```

Cette architecture garantit une **montée en complexité progressive** : on commence par établir les régularités moyennes (Partie 3), puis on teste leur hétérogénéité (Partie 4), avant d'élargir l'unité d'analyse au ménage (Partie 5).

### 5.6 Périmètre retenu

Il convient de préciser ce que ce travail modélise — et ce qu'il ne modélise pas.

**Ce que nous modélisons** :
- La probabilité d'**être** travailleur transfrontalier (variable binaire oui/non)
- Les déterminants **individuels** et **familiaux** de cette probabilité
- L'**hétérogénéité** de ces déterminants selon le genre
- La **corrélation** des décisions au sein des couples

**Ce que nous ne modélisons pas** :
- Le **choix du pays de destination** (Luxembourg vs Allemagne vs Suisse vs Belgique) — ce choix est considéré comme fortement endogène au lieu de résidence et aux réseaux préexistants
- La **dynamique temporelle** (entrée dans le statut frontalier, sortie, durée) — les données disponibles sont en coupe transversale
- L'**effet causal** des variables explicatives — notre analyse est associative, non causale

Cette délimitation du périmètre améliore la lisibilité des résultats et maintient la cohérence avec l'objectif principal : comprendre les mécanismes individuels et familiaux qui structurent la mobilité transfrontalière dans le Grand Est.

## 6. Données mobilisées

### 6.1 Sources et construction de la base

Ce travail exploite les données du **Recensement de la Population millésime 2022** produites par l'Institut National de la Statistique et des Études Économiques (INSEE). Deux fichiers de diffusion publique ont été mobilisés :

| Fichier | Contenu | Observations brutes | Variables |
|---------|---------|:-------------------:|:---------:|
| **MOBPRO** | Mobilités professionnelles (lieu de travail, mode de transport) | 553 997 | 32 |
| **INDCVI** | Caractéristiques individuelles détaillées (diplôme, CSP, situation familiale) | 1 326 580 | 94 |

*Note : Le millésime 2022 correspond aux données collectées sur la période 2018-2022, selon la méthodologie de recensement en continu de l'INSEE.*

Ces deux fichiers, bien que portant sur les mêmes individus, ne partagent **aucun identifiant individuel commun** dans leur version publique. Cette contrainte — inhérente à la protection du secret statistique — a imposé le recours à une méthode d'**appariement déterministe sur clé composite**, développée et validée dans la Partie 1 du projet.

### 6.2 Stratégie d'appariement

L'appariement repose sur la construction d'une clé composite combinant les variables communes aux deux fichiers :

$$\text{CLÉ} = f(\text{IRIS}, \text{SEXE}, \text{AGEREVQ}, \text{INATC}, \text{COUPLE}, \text{NPERR}, \ldots)$$

Une stratégie à **deux niveaux** a été mise en œuvre pour maximiser le taux de couverture tout en préservant la qualité des appariements :

| Niveau | Granularité géographique | Logique |
|:------:|--------------------------|---------|
| **Niveau 1** | IRIS (≈ 2 000 habitants) | Appariement fin, haute précision |
| **Niveau 2** | Canton-ou-ville | Récupération des communes rurales sans découpage IRIS |

Cette approche à deux niveaux était indispensable : un appariement limité au Niveau 1 aurait exclu **45 % des travailleurs ruraux**, créant un biais de sélection substantiel particulièrement problématique pour l'analyse du travail frontalier (les zones rurales frontalières étant fortement concernées par le phénomène).

### 6.3 Périmètre géographique et exclusion des Ardennes

Le périmètre géographique retenu couvre **9 des 10 départements** de la région Grand Est. Le département des **Ardennes (08)** a dû être exclu de l'analyse en raison d'une **anomalie de codage** détectée dans le fichier MOBPRO :

| Base | Département 08 (Ardennes) | Codes 80-85 |
|------|:-------------------------:|:-----------:|
| **MOBPRO** | ❌ **Absent** | 5 586 observations |
| **INDCVI** | ✅ 5 002 observations | ❌ Absents |

Les codes 80-85 présents dans MOBPRO ne correspondent à aucun département du Grand Est. Il s'agit vraisemblablement d'une **erreur de codage ou d'extraction** dans le fichier source diffusé par l'INSEE. Cette incohérence rend impossible l'appariement des individus ardennais entre les deux bases et impose leur exclusion.

**Conséquence** : Cette exclusion constitue une **limite du projet**, car elle réduit la couverture des zones frontalières avec la Belgique. Toutefois, plusieurs éléments atténuent l'impact de cette perte :

| Facteur atténuant | Explication |
|-------------------|-------------|
| **Faiblesse relative des flux** | Les Ardennes présentent un nombre de travailleurs frontaliers limité comparé aux départements mosellans ou alsaciens |
| **Destination unique** | Les flux sortants se dirigent quasi-exclusivement vers la Belgique, pays présentant des différentiels de salaires moins marqués que le Luxembourg ou la Suisse |
| **Spécificité territoriale** | Le tissu économique ardennais (industrie en déclin, ruralité prononcée) diffère des dynamiques frontalières observées ailleurs dans la région |

Les 9 départements retenus concentrent plus de **98 % des travailleurs frontaliers** de la région Grand Est, préservant la validité des conclusions pour l'essentiel du phénomène étudié.

### 6.4 Base finale

Après appariement, filtrage sur les actifs occupés et validation de la qualité des données, la base finale présente les caractéristiques suivantes :

| Caractéristique | Valeur |
|-----------------|:------:|
| **Observations** | 494 483 |
| **Variables** | 106 |
| **Départements couverts** | 9 (Moselle, Bas-Rhin, Haut-Rhin, Meurthe-et-Moselle, Meuse, Vosges, Marne, Aube, Haute-Marne) |
| **Taux de couverture** | 93,0 % |

La variable d'intérêt — le statut de travailleur transfrontalier — est construite à partir du lieu de travail déclaré :

| Statut | Effectif | Proportion |
|--------|:--------:|:----------:|
| **Travailleurs transfrontaliers** ($Y = 1$) | 44 264 | 8,95 % |
| **Non-transfrontaliers** ($Y = 0$) | 450 219 | 91,05 % |

Ce déséquilibre de classes (environ 9 % de positifs) est caractéristique des phénomènes de mobilité spécifique. Il ne constitue pas un obstacle à l'estimation économétrique — les modèles probit étant robustes aux déséquilibres modérés — mais impose une attention particulière dans l'interprétation des métriques de classification (privilégier l'AUC-ROC au taux de bonne classification).

### 6.5 Reconstruction des couples (Partie 5)

Pour l'estimation du probit bivarié, une base spécifique de **couples bi-actifs** a été reconstruite à partir de la variable LPRM (Lien à la Personne de Référence du Ménage) :

| Critère de sélection | Justification |
|----------------------|---------------|
| Exactement 1 individu avec LPRM = 1 | Une seule personne de référence par ménage |
| Exactement 1 individu avec LPRM = 2 | Un(e) seul(e) conjoint(e) identifié(e) |
| Les deux membres sont actifs occupés | Décision de travail transfrontalier possible pour les deux |
| Couple hétérosexuel | Interprétation H/F non ambiguë dans le cadre de l'analyse de genre |

Cette procédure aboutit à une base de **126 935 couples**, soit 253 870 individus appariés deux à deux. Le format adopté est un format **WIDE** (une ligne par couple), avec suffixes `_REF` et `_CONJ` pour les variables individuelles.

| Caractéristique | Valeur |
|-----------------|:------:|
| **Couples reconstruits** | 126 935 |
| **Score de qualité moyen** | 4,98 / 5 |
| **Cohérence déclarative** (COUPLE = 1) | 99,55 % |

## 7. Contribution de l'étude

### 7.1 Apports empiriques

Ce travail apporte plusieurs contributions à la connaissance du phénomène frontalier dans le Grand Est :

**Une quantification précise des déterminants individuels.** Au-delà des statistiques descriptives existantes, cette étude fournit des **effets marginaux** exprimés en points de pourcentage, permettant de hiérarchiser l'importance relative des différents facteurs. Par exemple, l'écart de probabilité entre un résident de Moselle et un résident de la Marne, toutes choses égales par ailleurs, peut être chiffré avec précision.

**Une analyse genrée des mécanismes.** L'introduction systématique d'interactions genre × covariables permet d'identifier les dimensions sur lesquelles hommes et femmes diffèrent significativement, et celles sur lesquelles les effets sont symétriques. Cette granularité est absente des analyses existantes.

**Une modélisation de la dimension conjugale.** À notre connaissance, aucune étude économétrique n'a, à ce jour, analysé la coordination des décisions de travail frontalier au sein des couples sur données françaises. L'estimation du paramètre de corrélation $\rho$ et le calcul des probabilités conditionnelles constituent un apport original.

### 7.2 Apports méthodologiques

Sur le plan méthodologique, ce travail illustre :

- La faisabilité d'un **appariement déterministe** entre fichiers INSEE sans identifiant commun, avec un taux de couverture de 93 %
- L'intérêt d'une **sélection de variables par apprentissage automatique** (Elastic Net) en amont de la modélisation économétrique, combinant objectivité statistique et arbitrages économiques
- La mise en œuvre d'un **probit bivarié** sur données françaises de grande dimension (> 125 000 couples), avec validation par tests de spécification adaptés aux grands échantillons

### 7.3 Portée et limites

Les résultats de cette étude présentent une **portée descriptive et associative**. Ils permettent d'identifier les facteurs corrélés au travail frontalier et de quantifier l'ampleur de ces associations. Ils ne permettent pas, en revanche, d'établir des relations causales : l'absence de variation exogène exploitable interdit toute interprétation en termes d'effet causal.

Cette limite, inhérente aux données en coupe transversale sans instrument, est pleinement assumée. L'objectif n'est pas de répondre à la question *« Que se passerait-il si l'on modifiait telle caractéristique ? »* (question causale), mais à la question *« Quelles caractéristiques sont associées au travail frontalier, et dans quelle mesure ? »* (question descriptive).

Les résultats peuvent néanmoins éclairer la réflexion des décideurs publics en identifiant les **profils** les plus concernés par le travail frontalier et les **mécanismes familiaux** qui structurent ces mobilités.

## 8. Organisation du mémoire

Ce mémoire est organisé en **cinq parties**, suivant une progression logique allant de la préparation des données jusqu'à la modélisation structurelle des décisions de ménage.

---

### Partie 1 — Préparation de la base de données

Cette première partie documente l'ensemble du travail de *data engineering* nécessaire à la constitution d'une base analytique exploitable. Elle détaille :

- L'exploration des fichiers sources MOBPRO et INDCVI
- Le diagnostic des incohérences et anomalies (variables communes, codages, valeurs manquantes)
- La stratégie d'appariement déterministe à deux niveaux
- La reconstruction des variables géographiques (DEPCOM, CANTVILLE)
- La validation de la qualité de l'appariement et l'analyse des biais résiduels

**Livrable** : Base finale de 494 483 observations × 106 variables, prête pour l'analyse économétrique.

---

### Partie 2 — Sélection des variables par apprentissage automatique

Cette deuxième partie présente la démarche de sélection des variables explicatives, combinant diagnostic statistique et arbitrages économiques. Elle comprend :

- L'analyse de multicolinéarité (matrices de Cramér, coefficients de Spearman, VIF)
- Le clustering hiérarchique pour identifier les « paniers » de variables redondantes
- Les arbitrages théoriques pour choisir entre variables corrélées
- La régularisation Elastic Net pour valider objectivement la sélection
- L'interprétation économique des coefficients pénalisés

**Livrable** : Ensemble de 17 variables conceptuelles (65 modalités après encodage) retenues pour la modélisation.

---

### Partie 3 — Modèle probit binaire : les déterminants individuels

Cette troisième partie constitue le cœur de l'analyse économétrique individuelle. Elle présente :

- L'estimation du modèle probit par maximum de vraisemblance
- Le calcul et l'interprétation des effets marginaux moyens (AME)
- Les diagnostics de qualité (pseudo-$R^2$, AIC, BIC, courbe ROC, calibration)
- Les tests de robustesse adaptés aux grands échantillons (Brier score, validation croisée)
- L'analyse économique des résultats par dimension thématique

**Livrable** : Quantification des effets des caractéristiques individuelles sur la probabilité de travail transfrontalier.

---

### Partie 4 — Modèle probit avec interactions : l'hétérogénéité selon le genre

Cette quatrième partie enrichit le modèle précédent en testant systématiquement l'hétérogénéité des effets entre hommes et femmes. Elle comprend :

- La spécification du modèle avec termes d'interaction genre × covariables
- Les tests de Wald par groupe thématique (parentalité, temps partiel, secteur, origine)
- Le calcul des effets marginaux différenciés pour les hommes et les femmes
- L'interprétation des mécanismes genrés (conciliation, segmentation, accès aux ressources)
- La comparaison avec le modèle sans interactions (test LR global)

**Livrable** : Identification des dimensions sur lesquelles les déterminants agissent différemment selon le genre.

---

### Partie 5 — Modèle probit bivarié : la coordination au sein des couples

Cette cinquième partie constitue l'apport le plus original du mémoire. Elle modélise simultanément les décisions des deux membres du couple et teste l'existence d'une coordination. Elle présente :

- La reconstruction de la base des couples bi-actifs (126 935 couples)
- La spécification du probit bivarié avec corrélation des résidus
- L'estimation du paramètre $\rho$ et les tests d'indépendance (Wald, LR)
- La comparaison des coefficients entre les deux équations (personne de référence vs conjoint)
- Le calcul des probabilités conditionnelles et de l'effet multiplicateur
- L'interprétation économique de la coordination intra-ménage

**Livrable** : Mesure de l'interdépendance des décisions conjugales et quantification de l'effet du statut du conjoint.

---

### Conclusion générale

La conclusion synthétise les principaux résultats, discute leurs implications pour la compréhension du phénomène frontalier, et identifie les limites de l'analyse ainsi que les pistes de prolongement.
