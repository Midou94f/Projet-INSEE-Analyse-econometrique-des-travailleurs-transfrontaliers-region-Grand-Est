# Conclusion générale

---

## 1. Synthèse de la démarche

Ce mémoire s'est attaché à comprendre les déterminants de la décision de travail transfrontalier dans le Grand Est, première région française concernée par ces mobilités professionnelles avec plus de 200 000 actifs traversant quotidiennement une frontière vers le Luxembourg, l'Allemagne, la Suisse ou la Belgique.

Trois questions de recherche complémentaires ont structuré l'analyse :

| Question | Niveau d'analyse | Modèle |
|----------|:----------------:|:------:|
| **Q1** — Quels facteurs individuels influencent la probabilité d'être transfrontalier ? | Individu | Probit binaire |
| **Q2** — Ces effets varient-ils selon le genre ? | Individu × Genre | Probit avec interactions |
| **Q3** — Les décisions des conjoints sont-elles coordonnées ? | Couple | Probit bivarié |

L'exploitation des données du Recensement 2022, après un travail d'appariement déterministe entre les fichiers MOBPRO et INDCVI (taux de couverture : 93 %), a permis de constituer une base de **494 483 actifs occupés** — dont 44 264 travailleurs transfrontaliers (8,95 %) — répartis sur 9 départements. La reconstruction d'une base de **126 935 couples bi-actifs** a ouvert la voie à l'analyse de la dimension conjugale, jusqu'alors inexplorée sur données françaises.

Les sections suivantes présentent les principaux résultats obtenus pour chacune des trois questions, avant d'en proposer une synthèse transversale et d'en discuter les implications, les forces, les limites et les prolongements possibles.

---

## 2. Réponse à la première question : les déterminants individuels

*Quels facteurs individuels influencent la probabilité d'être travailleur transfrontalier ?*

Le modèle probit binaire estimé dans la Partie 3 apporte une réponse structurée à cette question. Estimé par maximum de vraisemblance sur 494 483 observations, le modèle converge en 10 itérations et présente des indicateurs de qualité satisfaisants pour un modèle de choix discret sur données individuelles :

| Indicateur | Valeur | Interprétation |
|------------|:------:|----------------|
| Pseudo-R² de McFadden | **0,239** | Satisfaisant (seuil usuel : 0,20-0,40) |
| AUC-ROC | **0,849** | Excellent pouvoir discriminant |
| Brier Score | 0,068 | Réduction de 16,4 % vs modèle nul |
| Calibration (écart moyen) | 0,33 pp | Quasi-parfaite |
| Variables significatives (p < 1 %) | 56 / 65 | 86 % des variables |

La validation croisée stratifiée à 5 plis (AUC = 0,841 ± 0,004) confirme l'absence de sur-ajustement : le modèle généralise bien à des données non observées.

### 2.1 La géographie : déterminant dominant

Le résultat le plus saillant de l'analyse est la **prédominance écrasante de la géographie**. Les effets marginaux moyens (AME) des variables départementales — exprimés par rapport à la Marne, département non frontalier servant de référence — dominent tous les autres déterminants d'un ordre de grandeur :

| Département | AME (pp) | Interprétation |
|-------------|:--------:|----------------|
| **Moselle (57)** | **+30,6** | Proximité Luxembourg — effet maximal |
| **Haut-Rhin (68)** | **+26,3** | Proximité Suisse et Allemagne |
| **Meurthe-et-Moselle (54)** | **+24,4** | Proximité Luxembourg |
| **Meuse (55)** | **+19,2** | Position intermédiaire |
| **Bas-Rhin (67)** | **+18,5** | Proximité Allemagne |
| Vosges (88) | +6,7 | Effet modéré (relief, éloignement) |
| Aube (10) | +0,1 | Non significatif |
| Haute-Marne (52) | +0,1 | Non significatif |

Ces amplitudes sont considérables : résider en Moselle plutôt qu'en Marne augmente, *toutes choses égales par ailleurs*, la probabilité d'être transfrontalier de **31 points de pourcentage**. Dit autrement, à caractéristiques individuelles identiques (même âge, même diplôme, même profession, même situation familiale), un résident mosellan a une probabilité d'être frontalier supérieure de 31 points à celle d'un résident marnais.

Ce gradient spatial reflète fidèlement la structuration du phénomène autour des trois pôles d'attraction : le **Luxembourg** pour la Moselle et la Meurthe-et-Moselle, la **Suisse** pour le Haut-Rhin, et l'**Allemagne** pour le Bas-Rhin et dans une moindre mesure le Haut-Rhin. Les départements intérieurs (Aube, Haute-Marne), trop éloignés des frontières, ne présentent pas d'effet significatif.

### 2.2 Les réseaux transnationaux : deuxième facteur

Après la géographie, les variables liées à l'**origine et à la nationalité** constituent le deuxième bloc de déterminants par ordre d'importance :

| Variable | AME (pp) | Interprétation |
|----------|:--------:|----------------|
| Nationalité étrangère (INATC) | **+6,2** | Réseaux, capital linguistique, moindre aversion |
| Né à l'étranger (DNAI) | **+4,4** | Familiarité avec la mobilité internationale |
| Né dans un DOM-TOM | +0,8 | Effet faible (peu d'observations) |

Ces effets sont **cumulables** : un individu de nationalité étrangère né à l'étranger bénéficie d'un avantage de l'ordre de +10 pp par rapport à un Français né en métropole. Plusieurs mécanismes peuvent expliquer cette association :

- **Capital linguistique** : maîtrise de l'allemand, du luxembourgeois ou de l'anglais facilitant l'insertion sur les marchés du travail voisins
- **Réseaux informationnels** : connaissance des opportunités d'emploi via des liens familiaux ou communautaires transfrontaliers
- **Moindre aversion à la mobilité internationale** : familiarité avec les démarches administratives, absence de barrière psychologique à travailler « à l'étranger »
- **Proximité culturelle** : pour les ressortissants allemands, luxembourgeois, suisses ou belges résidant dans le Grand Est, le travail frontalier peut constituer un retour vers le pays d'origine

### 2.3 Le profil démographique : âge et genre

Les variables démographiques de base — âge et sexe — présentent des effets significatifs et économiquement interprétables.

**L'âge suit une trajectoire en U inversé.** La spécification quadratique (AGEREV + AGEREV²) est validée par le test du rapport de vraisemblance (χ² = 312, p < 0,001). L'effet marginal de l'âge est positif mais décroissant, avec un **pic estimé à 41 ans**. Avant cet âge, chaque année supplémentaire augmente la probabilité d'être frontalier (+0,41 pp par an en moyenne) ; après, l'effet s'inverse progressivement.

Cette forme en cloche reflète un double arbitrage :
- **Phase ascendante (avant 41 ans)** : accumulation d'expérience professionnelle, progression de carrière, maximisation des revenus — le travail frontalier, mieux rémunéré, devient attractif
- **Phase descendante (après 41 ans)** : pénibilité croissante des trajets quotidiens, ancrage territorial renforcé (propriété immobilière, enfants scolarisés, réseau social local), arbitrage en faveur de la qualité de vie

**Le genre structure les comportements.** Les femmes présentent une probabilité inférieure de **1,6 point de pourcentage** d'être travailleuses transfrontalières, toutes choses égales par ailleurs. Cet écart, modeste en apparence, masque une hétérogénéité substantielle qui sera explorée dans la Partie 4 : les interactions genre × covariables révèlent que les mécanismes à l'œuvre diffèrent profondément entre hommes et femmes.

### 2.4 La situation familiale : une contrainte croissante

La présence d'enfants dans le ménage exerce une **contrainte négative et croissante** sur la probabilité de travail transfrontalier :

| Nombre d'enfants | AME (pp) | Effet marginal par enfant |
|------------------|:--------:|:-------------------------:|
| 0 enfant | *Référence* | — |
| 1 enfant | **-0,9** | -0,9 pp |
| 2 enfants | **-1,2** | -0,3 pp |
| 3 enfants | **-2,4** | -1,2 pp |
| 4 enfants ou plus | **-2,7** | -0,3 pp |
| Enfants hors ménage | **+1,0** | — |

Deux observations émergent de ce tableau. Premièrement, l'effet est **monotone décroissant** : plus le nombre d'enfants augmente, plus la probabilité de travail frontalier diminue. Deuxièmement, la relation est **non linéaire** : l'effet marginal du troisième enfant (-1,2 pp) est quatre fois supérieur à celui du deuxième (-0,3 pp), suggérant l'existence d'un **seuil critique autour de trois enfants** au-delà duquel les contraintes logistiques deviennent rédhibitoires.

Ces contraintes sont de plusieurs ordres :
- **Incompatibilité temporelle** : les trajets frontaliers (souvent 1h à 1h30 par trajet) sont difficilement conciliables avec les horaires scolaires et les activités extrascolaires
- **Coûts de garde** : le recours à des modes de garde payants devient économiquement non viable au-delà de deux enfants, même avec les revenus frontaliers
- **Arbitrage conjugal** : dans les familles nombreuses, un parent tend à se spécialiser dans la logistique familiale tandis que l'autre maximise les revenus

L'effet positif associé aux enfants ayant quitté le ménage (+1,0 pp) confirme cette interprétation : une fois les contraintes parentales levées, la probabilité de travail frontalier augmente — effet de « libération ».

Le statut conjugal (vivre en couple) présente un effet positif mais modeste (+0,3 pp), qui sera éclairé par l'analyse de la coordination conjugale en Partie 5.

### 2.5 Les caractéristiques professionnelles : sélection sectorielle et statutaire

Le travail frontalier est fortement segmenté selon le **secteur d'activité** et le **statut d'emploi**.

**Secteur d'activité (NA5)** — par rapport au commerce et services (référence) :

| Secteur | AME (pp) | Interprétation |
|---------|:--------:|----------------|
| Agriculture (AZ) | **-12,0** | Activité terrienne non délocalisable |
| Administration publique (OQ) | **-8,3** | Secteur public français |
| Industrie (BE) | -0,2 | Non significatif |
| Construction (FZ) | +1,1 | Chantiers transfrontaliers |

L'agriculture et le secteur public constituent des **barrières structurelles** au travail frontalier. L'agriculture, par nature ancrée au territoire, ne permet pas la mobilité quotidienne vers l'étranger. Le secteur public français (fonction publique d'État, territoriale, hospitalière) n'a pas d'équivalent frontalier : un fonctionnaire français ne peut pas exercer au Luxembourg ou en Allemagne sans changer de statut.

**Statut d'emploi (EMPL)** — par rapport aux CDI/titulaires (référence) :

| Statut | AME (pp) | Interprétation |
|--------|:--------:|----------------|
| Emplois aidés | **-9,4** | Dispositifs nationaux non exportables |
| Indépendants | **-7,2** | Activité ancrée localement |
| Employeurs | -3,9 | Entreprise implantée en France |
| CDD | +0,7 | Flexibilité, entrée sur le marché |
| Intérim | -0,9 | Agences locales |

Les formes d'emploi **précaires ou spécifiques** (emplois aidés, indépendants) sont négativement associées au travail frontalier. Les emplois aidés correspondent à des dispositifs nationaux (contrats d'insertion, emplois d'avenir) qui n'existent pas à l'étranger. Les indépendants et employeurs, dont l'activité est par nature implantée sur un territoire, ne peuvent pas « traverser la frontière » quotidiennement.

**Catégorie socioprofessionnelle (GS)** — par rapport aux employés (référence) :

| CSP | AME (pp) | Interprétation |
|-----|:--------:|----------------|
| Ouvriers | **+1,2** | Industrie transfrontalière |
| Cadres | +0,8 | Tertiaire qualifié Luxembourg/Suisse |
| Professions intermédiaires | +0,4 | Position médiane |
| Artisans, commerçants | **-3,9** | Activité locale |
| Agriculteurs | -0,4 | Non significatif (effectif faible) |

Les **ouvriers** présentent la probabilité la plus élevée (+1,2 pp vs employés), reflétant l'importance de l'industrie dans les flux vers l'Allemagne et le Luxembourg. Les **cadres** bénéficient également d'un avantage (+0,8 pp), tiré par les emplois qualifiés du secteur financier luxembourgeois et des sièges sociaux suisses. Ce double profil — ouvriers qualifiés et cadres supérieurs — dessine un **marché du travail frontalier dual**.

### 2.6 Le capital humain : des rendements différenciés

Le niveau de diplôme présente des effets non linéaires par rapport au CAP-BEP (référence) :

| Diplôme | AME (pp) | Interprétation |
|---------|:--------:|----------------|
| Sans diplôme | -1,1 | Barrière à l'entrée |
| Brevet | -0,6 | Qualification insuffisante |
| Baccalauréat | +0,3 | Seuil d'accès |
| Bac+2 | +1,4 | Techniciens recherchés |
| Bac+3/4 | +1,9 | Professions intermédiaires |
| Bac+5 | **+2,8** | Cadres, finance, ingénierie |
| Doctorat | -6,6 | Carrières académiques françaises |
| Grandes écoles | **+3,1** | Élite professionnelle |

Le rendement du diplôme est **croissant jusqu'au Bac+5**, reflétant la prime aux qualifications sur les marchés du travail frontaliers (Luxembourg, Suisse). L'effet négatif du doctorat (-6,6 pp) s'explique par la spécificité des carrières académiques et de recherche, majoritairement ancrées dans le système universitaire français.

### 2.7 Variables potentiellement endogènes : associations significatives mais interprétation prudente

Plusieurs variables présentent des associations statistiquement significatives mais **potentiellement endogènes** — c'est-à-dire que la causalité pourrait être inverse (le travail frontalier causant la caractéristique observée) ou que les deux variables pourraient être causées par un facteur commun non observé.

**Statut d'occupation du logement** :
- Propriétaires : *référence*
- Locataires HLM : **-6,2 pp**
- Locataires privés : -0,7 pp
- Logés gratuitement : -0,8 pp

**Motorisation du ménage** :
- 1 voiture : *référence*
- 0 voiture : **-4,1 pp**
- 2 voitures : +1,7 pp
- 3+ voitures : +2,3 pp

Ces associations peuvent refléter trois mécanismes non exclusifs :
1. **Effet causal direct** : la voiture est nécessaire pour les trajets frontaliers (transports en commun peu développés), la propriété traduit un ancrage favorable
2. **Causalité inverse** : les revenus frontaliers permettent l'accession à la propriété et l'achat de véhicules
3. **Hétérogénéité non observée** : les locataires HLM et les ménages sans voiture cumulent des caractéristiques défavorables non captées par le modèle

**En l'absence d'instrument valide**, ces variables sont incluses comme contrôles mais leurs coefficients ne doivent pas être interprétés causalement. Les tests de robustesse (exclusion de ces variables) confirment la stabilité des coefficients des variables exogènes : l'effet de la Moselle varie de moins de 1 %, celui du sexe de 4,4 %.

## 3. Réponse à la deuxième question : l'hétérogénéité selon le genre

*Les effets des déterminants varient-ils selon le genre ?*

Le modèle probit binaire simple (Partie 3) a révélé un écart de 1,6 pp entre hommes et femmes dans la probabilité d'être travailleur transfrontalier. Cependant, ce coefficient unique suppose un effet homogène du genre et ne capture pas les mécanismes différenciés qui pourraient être à l'œuvre. La Partie 4 teste explicitement cette hétérogénéité en introduisant **45 termes d'interaction** entre la variable SEXE et les principales covariables du modèle.

### 3.1 Validation statistique de l'hétérogénéité

Avant d'interpréter les résultats, il convient d'établir que l'hétérogénéité observée est statistiquement significative et non le fruit du hasard. Trois tests convergent vers la même conclusion :

**Test de Wald global (toutes interactions)** :

| Métrique | Valeur |
|----------|:------:|
| Statistique de Wald | **926,52** |
| Degrés de liberté | 45 |
| P-value | < 10⁻¹⁶⁵ |

**Test du rapport de vraisemblance (LR)** :

| Métrique | Sans interactions | Avec interactions |
|----------|:-----------------:|:-----------------:|
| Log-vraisemblance | -113 380,76 | -112 913,73 |
| Paramètres | 66 | 111 |
| **Statistique LR** | | **934,07** |
| **P-value** | | ≈ 0 |

**Comparaison des critères d'information** :

| Critère | Sans interactions | Avec interactions | Verdict |

| AIC | 226 893,52 | 226 049,45 | ✓ Amélioré |

Les trois tests rejettent massivement l'hypothèse nulle d'homogénéité des effets. Le modèle avec interactions est **significativement meilleur** que le modèle sans interactions, et ce malgré l'ajout de 45 paramètres supplémentaires — les critères AIC et BIC, qui pénalisent la complexité, sont tous deux améliorés.

Sur les 45 interactions testées, **21 sont individuellement significatives** au seuil de 5 % (47 %). Les tests de Wald par groupe thématique confirment que 9 des 10 dimensions testées présentent des effets significativement différenciés :

Seul l'ancrage résidentiel (statut d'occupation du logement) ne présente pas d'effet significativement différencié entre hommes et femmes. Pour toutes les autres dimensions, **les déterminants n'agissent pas de la même manière selon le genre**.

### 3.2 Le temps partiel : seule variable changeant de signe

Le résultat le plus frappant de l'analyse concerne le **temps partiel**. C'est la seule variable du modèle dont l'effet marginal change de signe entre hommes et femmes :

| Variable | AME Hommes | AME Femmes | Δ (F-H) | IC 95 % |
|----------|:----------:|:----------:|:-------:|:-------:|
| Temps partiel (TP_2) | **-0,57 pp** | **+2,23 pp** | **+2,80 pp** | [+2,31 ; +3,29] |

Pour les hommes, travailler à temps partiel **réduit** légèrement la probabilité d'être transfrontalier (-0,57 pp). Pour les femmes, au contraire, le temps partiel l'**augmente** significativement (+2,23 pp). L'écart de 2,80 pp entre les deux genres est le plus important de toute l'analyse sur une variable individuelle.

**Interprétation économique** : Le temps partiel transfrontalier apparaît comme une **stratégie de conciliation spécifiquement féminine**. Face à la double contrainte des trajets longs (souvent 1h-1h30 par trajet) et des responsabilités domestiques (enfants, tâches ménagères), certaines femmes optent pour un emploi frontalier à temps partiel — solution leur permettant de bénéficier des salaires frontaliers tout en préservant du temps pour la sphère familiale.

Pour les hommes, le temps partiel n'a pas cette fonction de conciliation : il est davantage subi (emploi précaire, sous-emploi) et n'est pas associé au travail frontalier. Cette asymétrie révèle la **persistance des rôles de genre** dans l'organisation familiale : même lorsqu'elles travaillent à l'étranger, les femmes demeurent les principales responsables de la sphère domestique.

### 3.3 La parentalité : une pénalité différenciée au-delà de deux enfants

L'effet de la parentalité sur la probabilité de travail transfrontalier diffère selon le genre, mais de manière **non linéaire** :

| Nombre d'enfants | AME Hommes | AME Femmes | Δ (F-H) | Significatif ? |
|------------------|:----------:|:----------:|:-------:|:--------------:|
| 1 enfant | -1,31 pp | -0,59 pp | **+0,72 pp** | ✅ Oui |
| 2 enfants | -1,26 pp | -1,20 pp | +0,06 pp | ❌ Non |
| 3 enfants | -2,22 pp | -2,31 pp | -0,09 pp | ❌ Non |
| 4+ enfants | -2,28 pp | -2,78 pp | **-0,50 pp** | ⚠️ Tendance |
| Enfants partis | +0,56 pp | +1,69 pp | **+1,13 pp** | ✅ Oui |

Un pattern remarquable émerge de ce tableau :

- **1 enfant** : les femmes sont **moins pénalisées** que les hommes (+0,72 pp). Avec un seul enfant, les contraintes logistiques restent gérables et les femmes parviennent à concilier travail frontalier et parentalité — peut-être via le temps partiel.

- **2 enfants** : l'effet devient **similaire** pour les deux genres. Le deuxième enfant constitue un point d'équilibre où hommes et femmes font face aux mêmes contraintes.

- **3-4+ enfants** : la pénalité s'**intensifie pour les femmes**. Au-delà de deux enfants, les normes de genre reprennent le dessus : c'est sur les femmes que pèse prioritairement la charge de la logistique familiale, les excluant davantage du travail frontalier.

- **Enfants partis** : les femmes bénéficient **davantage** de cette libération (+1,13 pp vs hommes). Une fois les contraintes parentales levées, les femmes « rattrapent » les opportunités manquées pendant les années d'éducation des enfants.

**Seuil critique** : La charge familiale ne devient différenciée selon le genre qu'au-delà de **deux enfants**. Ce seuil correspond au passage d'une organisation familiale « gérable » à une situation où la spécialisation des rôles s'impose.

### 3.4 Le plafond de verre : diplômées et indépendantes pénalisées

Les femmes diplômées et les femmes entrepreneures bénéficient **moins des opportunités transfrontalières** que leurs homologues masculins :

**Statut d'indépendant** :

| Variable | AME Hommes | AME Femmes | Δ (F-H) | IC 95 % |
|----------|:----------:|:----------:|:-------:|:-------:|
| Indépendant (EMPL_21) | **+9,01 pp** | **+3,98 pp** | **-5,03 pp** | [-6,27 ; -3,79] |

L'écart de **-5,03 pp** est le plus important de toute l'analyse. Être indépendant augmente fortement la probabilité de travail transfrontalier pour les hommes (+9 pp), mais beaucoup moins pour les femmes (+4 pp). Les femmes entrepreneures semblent confrontées à des barrières spécifiques — accès aux marchés frontaliers, réseaux professionnels, contraintes de conciliation accrues pour les indépendantes — qui limitent leur capacité à exploiter les opportunités transfrontalières.

**Diplôme Bac+5** :

| Variable | AME Hommes | AME Femmes | Δ (F-H) | IC 95 % |
|----------|:----------:|:----------:|:-------:|:-------:|
| Bac+5 | **+3,82 pp** | **+2,12 pp** | **-1,69 pp** | [-2,47 ; -0,91] |

Le rendement du diplôme Bac+5 est significativement plus faible pour les femmes (-1,69 pp). À niveau de diplôme équivalent, les femmes très qualifiées accèdent moins aux emplois frontaliers — particulièrement les postes de cadres dans la finance luxembourgeoise ou l'industrie allemande. Ce résultat évoque l'existence d'un **plafond de verre transfrontalier**.

### 3.5 L'intersectionnalité positive : double avantage pour les femmes d'origine étrangère

Contrairement aux résultats précédents, certaines interactions révèlent un **avantage féminin**. C'est le cas de l'origine étrangère :

| Variable | AME Hommes | AME Femmes | Δ (F-H) | IC 95 % |
|----------|:----------:|:----------:|:-------:|:-------:|
| Né à l'étranger (DNAI_3) | +3,79 pp | **+6,27 pp** | **+2,48 pp** | [+1,78 ; +3,18] |

Les femmes nées à l'étranger bénéficient d'un effet presque deux fois supérieur à celui des hommes (+6,27 pp vs +3,79 pp). Ce résultat illustre une **intersectionnalité positive** : la combinaison genre × origine produit un avantage cumulatif plutôt qu'un cumul de désavantages.

Plusieurs mécanismes peuvent expliquer ce pattern :
- **Réseaux féminins transnationaux** : les femmes d'origine étrangère disposent de réseaux spécifiques (communautaires, familiaux) facilitant l'accès aux emplois frontaliers
- **Secteurs d'emploi** : certains secteurs à forte présence féminine (services à la personne, santé, hôtellerie-restauration) recrutent activement dans les pays frontaliers
- **Moindre aversion à la mobilité** : les femmes ayant déjà vécu une migration internationale présentent une familiarité avec la mobilité qui compense les freins liés au genre

### 3.6 L'accès inégal au véhicule : une ressource genrée

La motorisation du ménage présente des effets différenciés selon le genre :

| Motorisation | AME Hommes | AME Femmes | Δ (F-H) |
|--------------|:----------:|:----------:|:-------:|
| 0 voiture | -4,12 pp | -4,08 pp | +0,04 pp |
| 1 voiture | *Réf.* | *Réf.* | — |
| 2 voitures | +2,19 pp | +1,17 pp | **-1,02 pp** |
| 3+ voitures | +3,47 pp | +1,86 pp | **-1,61 pp** |

La multi-motorisation du ménage bénéficie **davantage aux hommes** qu'aux femmes. Lorsque le ménage possède deux voitures, l'effet sur la probabilité de travail frontalier est de +2,19 pp pour les hommes mais seulement +1,17 pp pour les femmes (-1,02 pp d'écart). Avec trois voitures ou plus, l'écart se creuse encore (-1,61 pp).

**Interprétation** : Ce résultat suggère une **répartition genrée de l'usage automobile** au sein du ménage. Même lorsque plusieurs véhicules sont disponibles, c'est prioritairement l'homme qui utilise la voiture pour les trajets domicile-travail (longs trajets frontaliers), tandis que la femme utilise le second véhicule pour des trajets plus courts (accompagnement des enfants, courses, emploi local). La voiture, ressource partagée du ménage, n'est pas également accessible aux deux conjoints pour le travail frontalier.

### 3.7 Le secteur d'activité : des segmentations différentes

Les effets sectoriels diffèrent significativement selon le genre :

| Secteur | AME Hommes | AME Femmes | Δ (F-H) |
|---------|:----------:|:----------:|:-------:|

| Construction (FZ) | +1,45 pp | +0,42 pp | -1,03 pp |
| Administration (OQ) | -9,12 pp | -7,48 pp | **+1,64 pp** |

Le secteur public pénalise **moins les femmes** que les hommes (+1,64 pp d'écart). L'administration publique, bien que fermée au travail frontalier, constitue pour les femmes un employeur offrant des conditions de travail compatibles avec la vie familiale (horaires, congés, stabilité). Le « coût d'opportunité » de ne pas être frontalière est donc moindre pour les femmes que pour les hommes.

### 3.8 Synthèse : cinq mécanismes genrés identifiés

L'analyse des interactions permet d'identifier **cinq mécanismes genrés** structurant l'accès au travail transfrontalier :

| # | Mécanisme | Résultat clé | Interprétation |
|:-:|-----------|--------------|----------------|
| 1 | **Temps partiel = stratégie féminine** | H : -0,6 pp / F : +2,2 pp | Seule variable changeant de signe — compromis entre revenus frontaliers et contraintes domestiques |
| 2 | **Plafond de verre** | Indépendantes : -5,0 pp vs H | Les femmes entrepreneures et diplômées bénéficient moins des opportunités transfrontalières |
| 3 | **Pénalité parentale différenciée** | Écart F-H croissant > 2 enfants | Les normes de genre s'intensifient avec la charge familiale |
| 4 | **Intersectionnalité positive** | Femmes nées étranger : +2,5 pp vs H | Double avantage genre × origine |
| 5 | **Accès inégal au véhicule** | Multi-motorisation : -1,6 pp pour F | Répartition genrée de l'usage automobile |

Ces résultats démontrent que le travail transfrontalier **n'est pas un phénomène neutre du point de vue du genre**. Les facteurs qui favorisent ou freinent la mobilité transfrontalière n'agissent pas de la même manière pour les hommes et les femmes. Cette hétérogénéité substantielle — validée statistiquement par les tests de Wald et du rapport de vraisemblance — invite à examiner comment ces arbitrages se jouent au niveau du couple, objet de la Partie 5.

## 4. Réponse à la troisième question : la coordination conjugale

*Les décisions des conjoints sont-elles coordonnées ?*

La Partie 4 a révélé que les déterminants du travail transfrontalier diffèrent selon le genre, avec notamment des effets différenciés du temps partiel, de la parentalité et de l'accès au véhicule. Ces résultats suggèrent l'existence d'arbitrages au sein des ménages. La Partie 5 teste explicitement cette hypothèse en modélisant **simultanément** les décisions des deux membres du couple via un probit bivarié.

### 4.1 Construction de la base des couples

L'analyse de la coordination conjugale requiert une base de données spécifique où l'unité d'observation n'est plus l'individu mais le **couple**. Cette base a été reconstruite à partir de la variable LPRM (Lien à la Personne de Référence du Ménage) selon les critères suivants :

| Critère | Justification |
|---------|---------------|
| Exactement 1 individu avec LPRM = 1 | Une seule personne de référence par ménage |
| Exactement 1 individu avec LPRM = 2 | Un(e) seul(e) conjoint(e) identifié(e) |
| Les deux membres sont actifs occupés | Décision de travail transfrontalier possible pour les deux |
| Couple hétérosexuel | Interprétation H/F non ambiguë |

Cette procédure aboutit à une base de **126 935 couples**, soit 253 870 individus appariés deux à deux. Le format adopté est un format WIDE (une ligne par couple), avec des variables individuelles suffixées `_REF` (personne de référence, majoritairement l'homme) et `_CONJ` (conjoint, majoritairement la femme), et des variables de ménage communes (département, motorisation, nombre d'enfants).

| Caractéristique | Valeur |
|-----------------|:------:|
| Couples reconstruits | 126 935 |
| Score de qualité moyen | 4,98 / 5 |
| Cohérence déclarative (COUPLE = 1) | 99,55 % |

### 4.2 Statistiques descriptives : une corrélation brute élevée

Avant même l'estimation du modèle, les statistiques descriptives révèlent une **association forte** entre les décisions des deux conjoints :

|  | Femme non-transf. | Femme transf. | Total |
|--|:-----------------:|:-------------:|:-----:|
| **Homme non-transf.** | 108 827 (85,7 %) | 5 030 (4,0 %) | 113 857 (89,7 %) |
| **Homme transf.** | 7 003 (5,5 %) | 6 075 (4,8 %) | 13 078 (10,3 %) |
| **Total** | 115 830 (91,2 %) | 11 105 (8,8 %) | 126 935 (100 %) |

Sous hypothèse d'indépendance, la probabilité qu'un couple soit bi-transfrontalier (les deux conjoints travaillant à l'étranger) serait de :

$$P(1,1)_{\text{indép.}} = P(\text{REF}=1) \times P(\text{CONJ}=1) = 10{,}3\% \times 8{,}8\% = 0{,}91\%$$

Or, la fréquence observée est de **4,79 %**, soit **5,3 fois plus** que sous indépendance. Le coefficient Phi brut (corrélation entre variables binaires) atteint **0,452** — valeur remarquablement élevée pour des variables dichotomiques (> 0,3 = forte association).

Cette corrélation brute combine trois composantes :
1. L'**homogamie observée** : les conjoints partagent des caractéristiques similaires (diplôme, CSP, origine)
2. La **localisation commune** : les deux membres résident dans le même département
3. La **coordination inobservée** : facteurs non captés par les variables du modèle

Le probit bivarié permet d'isoler la troisième composante en contrôlant les deux premières.

### 4.3 Le modèle probit bivarié : spécification

Le probit bivarié modélise deux variables latentes simultanées :

$$\begin{cases} Y_{REF}^* = X_{REF}'\beta_{REF} + Z'\gamma_{REF} + \varepsilon_{REF} \\ Y_{CONJ}^* = X_{CONJ}'\beta_{CONJ} + Z'\gamma_{CONJ} + \varepsilon_{CONJ} \end{cases}$$

avec une distribution jointe des résidus :

$$\begin{pmatrix} \varepsilon_{REF} \\ \varepsilon_{CONJ} \end{pmatrix} \sim \mathcal{N}_2 \left( \begin{pmatrix} 0 \\ 0 \end{pmatrix}, \begin{pmatrix} 1 & \rho \\ \rho & 1 \end{pmatrix} \right)$$

Le paramètre clé est **ρ**, coefficient de corrélation entre les résidus des deux équations. Il mesure l'association entre les décisions des deux conjoints *après contrôle* de l'ensemble des caractéristiques observables (département, diplôme, CSP, âge, secteur, etc.).

| Valeur de ρ | Interprétation | Mécanisme économique |
|:-----------:|----------------|----------------------|
| ρ > 0 | Corrélation positive | **Coordination** : les deux conjoints tendent à être transfrontaliers ensemble |
| ρ < 0 | Corrélation négative | **Substitution** : si l'un est transfrontalier, l'autre tend à ne pas l'être |
| ρ = 0 | Indépendance | Pas de coordination — décisions individuelles |

### 4.4 Résultats d'estimation : ρ = 0,64

L'estimation par maximum de vraisemblance converge après optimisation L-BFGS-B. Les résultats sont sans ambiguïté :

| Paramètre | Estimation | Écart-type | IC 95 % |
|-----------|:----------:|:----------:|:-------:|
| **ρ** | **0,6423** | 0,0117 | [0,619 ; 0,665] |

Le coefficient de corrélation estimé est **fortement positif** (ρ = 0,64) et très précisément estimé (écart-type = 0,012). L'intervalle de confiance à 95 % exclut largement la valeur zéro.

**Tests d'hypothèse H₀ : ρ = 0** :

| Test | Statistique | P-value | Conclusion |
|------|:-----------:|:-------:|------------|
| Ratio de vraisemblance (LR) | χ² = **7 541,06** | < 10⁻¹⁰ | H₀ rejetée *** |
| Wald | z = **54,67** | < 10⁻¹⁰ | H₀ rejetée *** |

Les deux tests rejettent **massivement** l'hypothèse d'indépendance. La statistique LR de 7 541 correspond à une amélioration de la log-vraisemblance de 3 770 points par rapport au modèle à résidus indépendants — gain considérable pour un seul paramètre.

**Comparaison des log-vraisemblances** :

| Modèle | Log-vraisemblance |
|--------|:-----------------:|
| Probits indépendants (ρ = 0) | -60 014,96 |
| Probit bivarié (ρ libre) | -56 244,42 |
| **Amélioration** | **+3 770,53** |

### 4.5 Interprétation : la corrélation résiduelle

Le coefficient ρ = 0,64 mesure la corrélation entre les résidus des deux équations, c'est-à-dire l'association entre les décisions des conjoints **après avoir contrôlé** :

- Le département de résidence (commun aux deux conjoints)
- Le niveau de diplôme de chacun
- La catégorie socioprofessionnelle de chacun
- Le secteur d'activité de chacun
- L'âge de chacun
- Le type de contrat de chacun
- La nationalité et le lieu de naissance de chacun
- Le nombre d'enfants du ménage
- La motorisation du ménage
- Le statut d'occupation du logement

Malgré ce contrôle extensif, il subsiste une **corrélation forte et positive** entre les propensions au travail transfrontalier des deux conjoints. Cette corrélation résiduelle capture des facteurs non observés communs au couple :

**1. Préférences partagées non observées**
- Tolérance aux trajets longs
- Appétence pour le différentiel salarial
- Ouverture internationale, bilinguisme
- Aversion au risque similaire

**2. Effets de réseau et transmission d'information**
- Un conjoint déjà transfrontalier réduit les coûts d'entrée pour l'autre
- Information sur les opportunités d'emploi
- Connaissance des procédures administratives
- Réseau professionnel mobilisable

**3. Coordination stratégique explicite**
- Covoiturage (économie de temps et d'argent)
- Synchronisation des horaires
- Organisation de la garde d'enfants
- Mutualisation des réseaux avec d'autres couples frontaliers

**4. Localisation résidentielle fine**
- Au-delà du département (contrôlé), la localisation infra-communale détermine conjointement les opportunités
- Un couple installé à 5 km de la frontière luxembourgeoise fait face aux mêmes opportunités pour les deux membres

**Fait remarquable** : ρ (0,64) > Phi brut (0,45). La corrélation résiduelle est **supérieure** à la corrélation brute. Cela signifie que les variables de contrôle (département, diplôme, etc.) **masquaient partiellement** la vraie corrélation entre les décisions. L'homogamie observée (conjoints de même diplôme, même CSP) tend à réduire la corrélation apparente ; une fois contrôlée, la coordination sous-jacente apparaît plus forte.

### 4.6 Probabilités conditionnelles : l'effet multiplicateur ×4,7

La traduction opérationnelle du paramètre ρ passe par le calcul des **probabilités conditionnelles**. Quelle est la probabilité qu'un individu soit transfrontalier sachant que son conjoint l'est (ou ne l'est pas) ?

**Probabilités marginales** :
- P(REF transfrontalier) = 10,30 %
- P(CONJ transfrontalier) = 8,75 %

**Probabilités conditionnelles** :

| Condition | Probabilité | vs Marginale |
|-----------|:-----------:|:------------:|
| P(REF=1 \| CONJ=1) | **47,50 %** | +37,2 pp |
| P(REF=1 \| CONJ=0) | 6,59 % | -3,7 pp |
| P(CONJ=1 \| REF=1) | **47,66 %** | +38,9 pp |
| P(CONJ=1 \| REF=0) | 6,55 % | -2,2 pp |

Le résultat est saisissant :

> **Un individu dont le conjoint est transfrontalier a près d'une chance sur deux (47,5 %) d'être lui-même transfrontalier**, contre seulement 6,6 % si le conjoint ne l'est pas.

**Effet multiplicateur** :

$$\frac{P(\text{transf.} | \text{conjoint transf.})}{P(\text{transf.} | \text{conjoint non-transf.})} = \frac{47{,}5\%}{6{,}6\%} = \mathbf{4{,}7}$$

Avoir un conjoint transfrontalier **multiplie par 4,7** la probabilité de l'être soi-même. En termes absolus, l'écart est de **41 points de pourcentage** — effet massif.

**Symétrie parfaite** : Les effets sont quasi-identiques dans les deux sens (REF → CONJ et CONJ → REF), confirmant que la coordination est **bidirectionnelle**. Il n'y a pas de « leader » et de « suiveur » : les deux conjoints s'influencent mutuellement.

### 4.7 Odds ratio : les couples bi-transfrontaliers très surreprésentés

L'odds ratio mesure la surreprésentation des couples bi-transfrontaliers par rapport à l'hypothèse d'indépendance :

$$\text{Odds ratio} = \frac{P(1,1) \times P(0,0)}{P(1,0) \times P(0,1)} = \frac{4{,}91\% \times 85{,}86\%}{5{,}39\% \times 3{,}84\%} = \mathbf{12{,}8}$$

Les couples bi-transfrontaliers sont **12,8 fois plus fréquents** qu'ils ne le seraient si les décisions des conjoints étaient indépendantes. Cette valeur, très élevée, confirme que les couples où les deux membres travaillent à l'étranger constituent une **catégorie spécifique** de ménages — non pas une coïncidence statistique mais le résultat d'une logique de coordination.

### 4.8 Comparaison des coefficients entre équations

Le probit bivarié estime des coefficients séparés pour chaque équation (β_REF et β_CONJ). La comparaison de ces coefficients révèle si les déterminants agissent de la même manière pour les deux conjoints.

**Corrélation globale** : r(β_REF, β_CONJ) = **0,92**

Cette corrélation très élevée indique que les **mêmes facteurs** influencent la probabilité de travail transfrontalier pour les deux membres du couple. La géographie, la nationalité, le diplôme, le secteur d'activité jouent de manière similaire pour REF et CONJ.

Être femme pénalise **moins la personne de référence** que le conjoint. Une femme désignée comme personne de référence du ménage est en position professionnelle plus affirmée — ce qui réduit l'écart de genre.

Dans le cadre bivarié, les effets marginaux moyens (AME) intègrent la corrélation ρ entre les deux équations. Les principaux déterminants sont cohérents entre les deux équations :

### 4.10 Synthèse : le ménage comme unité d'analyse pertinente

L'estimation du probit bivarié établit de manière robuste l'existence d'une **coordination intra-ménage** dans les décisions de travail transfrontalier. Les résultats clés sont résumés dans le tableau suivant :

## 5. Synthèse transversale : faits stylisés et articulation des modèles

Les trois modèles probit estimés dans ce mémoire — binaire simple, avec interactions de genre, bivarié pour couples — forment un ensemble cohérent permettant de dresser un portrait complet des déterminants du travail transfrontalier dans le Grand Est. Cette section propose une synthèse transversale des résultats, organisée autour de sept faits stylisés quantifiés et d'une réflexion sur l'articulation des trois niveaux d'analyse.

### 5.1 Sept faits stylisés

**Fait 1 — La géographie structure massivement le phénomène**

Les effets départementaux dominent tous les autres déterminants d'un ordre de grandeur. Résider en Moselle plutôt qu'en Marne augmente la probabilité d'être transfrontalier de **30,6 points de pourcentage** ; le Haut-Rhin ajoute 26,3 pp, la Meurthe-et-Moselle 24,4 pp. Ces amplitudes, sans commune mesure avec celles des autres variables (généralement < 5 pp), reflètent la polarisation du phénomène autour des trois bassins d'emploi frontaliers : Luxembourg (Moselle, Meurthe-et-Moselle), Suisse (Haut-Rhin), Allemagne (Bas-Rhin, Haut-Rhin).

Le gradient spatial est cohérent avec la géographie économique de la région : les départements intérieurs (Aube, Haute-Marne), trop éloignés des frontières, ne présentent pas d'effet significatif. La distance physique à la frontière — non directement mesurée mais captée par les indicatrices départementales — constitue le **premier filtre** de la décision de travail transfrontalier.

**Fait 2 — Les réseaux transnationaux facilitent l'accès**

Après la géographie, les variables liées à l'origine et à la nationalité constituent le deuxième bloc de déterminants. La nationalité étrangère augmente la probabilité de **6,2 pp**, la naissance à l'étranger de **4,4 pp**, ces effets étant cumulables. Un individu de nationalité étrangère né hors de France bénéficie d'un avantage de l'ordre de **+10 pp** par rapport à un Français né en métropole.

Ces effets traduisent l'importance des réseaux transnationaux — capital linguistique, réseaux informationnels, familiarité avec la mobilité internationale — dans l'accès au marché du travail frontalier. Le travail transfrontalier n'est pas seulement une question de localisation géographique ; il suppose également une **proximité culturelle et sociale** avec les pays voisins.

**Fait 3 — Le secteur public et l'agriculture sont structurellement exclus**

Certains secteurs d'activité constituent des **barrières structurelles** au travail transfrontalier. L'agriculture (-12,0 pp) et le secteur public (-8,3 pp) présentent des effets négatifs massifs, reflétant l'impossibilité technique ou institutionnelle d'exercer ces activités à l'étranger. Un agriculteur ne peut pas cultiver ses terres au Luxembourg ; un fonctionnaire français ne peut pas exercer dans l'administration allemande.

Ces barrières sectorielles dessinent la géographie sociale du travail frontalier : une population **exclue de facto** (agriculteurs, fonctionnaires) coexiste avec une population **éligible** (industrie, services privés, construction). Les politiques publiques d'accompagnement du travail frontalier doivent intégrer cette segmentation.

**Fait 4 — La contrainte familiale est croissante et non linéaire**

La présence d'enfants dans le ménage réduit la probabilité de travail transfrontalier, avec un effet **monotone décroissant** : de -0,9 pp pour un enfant à -2,7 pp pour quatre enfants ou plus. Cette relation est non linéaire : l'effet marginal du troisième enfant (-1,2 pp) est quatre fois supérieur à celui du deuxième (-0,3 pp), suggérant un **seuil critique autour de trois enfants** au-delà duquel les contraintes logistiques deviennent rédhibitoires.

La levée des contraintes parentales (enfants ayant quitté le ménage) s'accompagne d'un effet positif (+1,0 pp), confirmant l'interprétation en termes de contrainte temporaire. La parentalité **contraint mais n'exclut pas définitivement** : une fois les enfants partis, la probabilité de travail frontalier augmente.

**Fait 5 — Le genre structure les comportements de manière substantielle**

L'analyse des interactions révèle que **47 % des termes d'interaction** genre × covariables sont statistiquement significatifs. Le travail transfrontalier n'est pas un phénomène neutre du point de vue du genre : les mêmes caractéristiques n'ont pas les mêmes effets pour les hommes et les femmes.

Le résultat le plus frappant concerne le **temps partiel**, seule variable dont l'effet change de signe entre genres : négatif pour les hommes (-0,6 pp), positif pour les femmes (+2,2 pp). Le temps partiel transfrontalier apparaît comme une **stratégie de conciliation spécifiquement féminine**, permettant de bénéficier des salaires frontaliers tout en préservant du temps pour la sphère domestique.

Les femmes diplômées et entrepreneures font face à un **plafond de verre transfrontalier** : le rendement du Bac+5 est inférieur de 1,7 pp pour les femmes, celui du statut d'indépendant de 5,0 pp. La pénalité parentale s'intensifie pour les femmes au-delà de deux enfants. Ces résultats convergent vers un constat : les normes de genre et la division sexuée du travail domestique **contraignent différemment** l'accès des hommes et des femmes au travail frontalier.

**Fait 6 — Les décisions conjugales sont fortement coordonnées**

Le probit bivarié établit l'existence d'une **coordination intra-ménage** robuste. Le coefficient de corrélation résiduelle ρ = 0,64 (p < 10⁻¹⁰) traduit une association forte entre les décisions des conjoints, *après contrôle* de l'ensemble des caractéristiques observables.

La traduction opérationnelle est saisissante : avoir un conjoint transfrontalier **multiplie par 4,7** la probabilité de l'être soi-même (47,5 % vs 6,6 %). Les couples bi-transfrontaliers sont **12,8 fois plus fréquents** qu'ils ne le seraient sous hypothèse d'indépendance. Cette surreprésentation massive confirme que les couples où les deux membres travaillent à l'étranger constituent une **catégorie spécifique** de ménages, résultat d'une logique de coordination et non d'une coïncidence statistique.

**Fait 7 — Les trois modèles sont statistiquement robustes et convergents**

Les trois modèles présentent des indicateurs de qualité satisfaisants et convergents :

| Indicateur | Partie 3 | Partie 4 | Partie 5 |
|------------|:--------:|:--------:|:--------:|
| Pseudo-R² | 0,239 | 0,242 | — |
| AUC-ROC | 0,849 | 0,851 | — |
| Log-vraisemblance | -113 381 | -112 914 | -56 244 |
| Convergence | 10 itérations | 10 itérations | Oui |
| Test global (Wald/LR) | χ² = 146 851 | χ² = 934 (interactions) | χ² = 7 541 (ρ) |

Les tests de robustesse confirment la stabilité des résultats : probit vs logit (corrélation 0,995), AME vs MEM (corrélation 1,00), inclusion/exclusion des variables potentiellement endogènes (variation < 5 %). Les trois niveaux d'analyse — individuel, genré, conjugal — produisent des résultats cohérents et complémentaires.

### 5.2 Articulation des trois niveaux d'analyse

Les trois modèles estimés forment une **progression logique** permettant une compréhension de plus en plus fine du phénomène :

```
┌─────────────────────────────────────────────────────────────────────┐
│  NIVEAU 1 — INDIVIDU (Partie 3)                                     │
│  Question : Qui devient transfrontalier ?                           │
│  Réponse : La géographie domine, suivie des réseaux transnationaux  │
│            et du profil socioprofessionnel                          │
└─────────────────────────────────────────────────────────────────────┘
                                  ↓
         Mais ces effets sont-ils homogènes ?
                                  ↓
┌─────────────────────────────────────────────────────────────────────┐
│  NIVEAU 2 — INDIVIDU × GENRE (Partie 4)                             │
│  Question : Les hommes et les femmes sont-ils égaux face au         │
│             travail frontalier ?                                    │
│  Réponse : Non — 47 % des effets diffèrent significativement        │
│            Temps partiel = stratégie féminine, plafond de verre     │
└─────────────────────────────────────────────────────────────────────┘
                                  ↓
         Mais ces arbitrages genrés se font dans quel cadre ?
                                  ↓
┌─────────────────────────────────────────────────────────────────────┐
│  NIVEAU 3 — COUPLE (Partie 5)                                       │
│  Question : Les conjoints se coordonnent-ils ?                      │
│  Réponse : Oui — ρ = 0,64, effet multiplicateur ×4,7                │
│            Le ménage est l'unité d'analyse pertinente               │
└─────────────────────────────────────────────────────────────────────┘
```

Cette progression révèle que la décision de travail transfrontalier **ne peut pas être comprise à l'échelle de l'individu isolé**. Le niveau individuel (Partie 3) identifie les déterminants moyens mais masque l'hétérogénéité de genre (Partie 4). L'hétérogénéité de genre, à son tour, ne prend sens que replacée dans le contexte des arbitrages conjugaux (Partie 5).

### 5.3 Une lecture unifiée : le travail frontalier comme décision de ménage

La synthèse des trois niveaux d'analyse conduit à une lecture unifiée : le travail transfrontalier est une **décision de ménage**, prise collectivement par les deux conjoints en fonction de contraintes géographiques, sectorielles, familiales et de genre.

**Schéma décisionnel stylisé** :

1. **Filtre géographique** : La localisation résidentielle détermine l'éligibilité. Seuls les ménages situés dans les départements frontaliers (Moselle, Haut-Rhin, Meurthe-et-Moselle, etc.) ont accès significatif au marché du travail frontalier.

2. **Filtre sectoriel** : Parmi les ménages éligibles géographiquement, seuls ceux dont au moins un membre travaille dans le secteur privé (hors agriculture) peuvent envisager le travail frontalier.

3. **Arbitrage familial** : Pour les ménages passant les deux premiers filtres, la présence d'enfants impose des contraintes logistiques. Ces contraintes pèsent différemment sur les hommes et les femmes, conduisant à des stratégies différenciées (temps partiel féminin, spécialisation d'un conjoint).

4. **Coordination conjugale** : La décision finale est prise conjointement. Si un membre du couple devient transfrontalier, la probabilité que l'autre le devienne également est multipliée par 4,7 — effet de réseau, coordination logistique, préférences partagées.

Cette lecture explique pourquoi les couples bi-transfrontaliers sont si fréquents (12,8 fois plus qu'attendu) : ce ne sont pas deux décisions individuelles indépendantes, mais une **stratégie de ménage** visant à maximiser les revenus tout en gérant les contraintes familiales.

## 6. Implications pour les politiques publiques

Les résultats de cette étude, bien que de portée descriptive et non causale, éclairent la réflexion des décideurs publics sur plusieurs dimensions. Les quelque 200 000 travailleurs frontaliers du Grand Est — et leurs familles — constituent une population aux caractéristiques et aux besoins spécifiques, que les politiques d'aménagement du territoire, de transport, d'emploi et de services aux familles gagneraient à mieux prendre en compte.

### 6.1 Cibler les ménages plutôt que les individus

Le résultat central de cette étude — la corrélation résiduelle ρ = 0,64 et l'effet multiplicateur ×4,7 — invite à repenser le ciblage des politiques d'accompagnement du travail frontalier.

**Constat** : Les décisions de travail transfrontalier ne sont pas prises individuellement mais dans une logique de ménage. Lorsqu'un membre d'un couple devient transfrontalier, la probabilité que son conjoint le devienne également passe de 6,6 % à 47,5 %. Les couples bi-transfrontaliers, 12,8 fois plus fréquents qu'attendu sous indépendance, ne sont pas une anomalie statistique mais le résultat d'une coordination conjugale.

**Implication** : Toute politique facilitant l'entrée sur le marché du travail frontalier produira des **effets démultiplicateurs** au niveau du ménage. Accompagner un individu, c'est potentiellement faciliter l'accès de son conjoint. Les dispositifs d'information, de formation linguistique, d'aide à la recherche d'emploi ou de reconnaissance des diplômes gagneraient à intégrer cette dimension familiale — par exemple en proposant des sessions ouvertes aux couples, en communiquant sur les avantages d'une stratégie coordonnée, ou en facilitant les démarches administratives conjointes.

**Recommandation** : Les guichets uniques d'accompagnement des travailleurs frontaliers (Maisons de l'emploi transfrontalières, EURES, etc.) pourraient systématiser une approche « ménage » dans leurs services d'orientation et d'accompagnement.

### 6.2 Adapter les infrastructures de transport aux flux de couples

La concentration des couples bi-transfrontaliers — près de 5 % des couples bi-actifs des départements frontaliers — justifie des investissements ciblés en matière d'infrastructures de transport.

**Constat** : Les couples où les deux conjoints travaillent à l'étranger font face à des contraintes logistiques spécifiques. Ils doivent organiser deux trajets quotidiens vers la frontière, souvent vers le même bassin d'emploi (Luxembourg, Bâle, Sarrebruck). Le covoiturage intra-couple est une solution naturelle mais suppose une synchronisation des horaires et un accès à un véhicule adapté.

**Implication** : Les infrastructures de transport transfrontalières doivent être dimensionnées pour absorber des flux de couples et non seulement d'individus isolés. Cela concerne :

| Infrastructure | Adaptation recommandée |
|----------------|------------------------|
| **Parkings relais** | Dimensionnement pour des flux importants aux heures de pointe (6h-8h, 17h-19h), avec bornes de recharge électrique |
| **Transports en commun** | Lignes transfrontalières à haute fréquence permettant une flexibilité horaire (départs toutes les 10-15 min aux heures de pointe) |
| **Covoiturage organisé** | Plateformes dédiées aux trajets frontaliers, intégrant la dimension couple (matching par zone de résidence et horaires compatibles) |
| **Pistes cyclables** | Pour les trajets de rabattement vers les gares et parkings relais |

**Recommandation** : Les schémas régionaux de transport (SRADDET) et les projets d'investissement transfrontaliers (Eurodistricts, GECT) devraient intégrer explicitement la dimension « couples bi-frontaliers » dans leurs études de besoins et de dimensionnement.

### 6.3 Développer les services aux familles compatibles avec les trajets longs

L'analyse de la parentalité révèle une contrainte croissante et non linéaire : la probabilité de travail frontalier diminue significativement au-delà de deux enfants (-2,4 pp pour trois enfants, -2,7 pp pour quatre ou plus). Cette contrainte pèse différemment sur les hommes et les femmes, conduisant certaines femmes à adopter le temps partiel comme stratégie de conciliation.

**Constat** : Les trajets frontaliers (souvent 1h à 1h30 par trajet, soit 2h à 3h quotidiennes) sont difficilement compatibles avec les horaires scolaires standards (8h30-16h30) et les activités extrascolaires. Les familles frontalières doivent trouver des solutions de garde avant et après l'école, ce qui représente un coût financier et organisationnel significatif.

**Implication** : Le développement de services aux familles adaptés aux contraintes des travailleurs frontaliers pourrait lever une partie de ces freins. Plusieurs pistes méritent attention :

| Service | Adaptation recommandée |
|---------|------------------------|
| **Crèches et haltes-garderies** | Horaires élargis (6h-20h) dans les communes frontalières à forte concentration de travailleurs frontaliers |
| **Accueil périscolaire** | Garderies scolaires ouvrant dès 6h30 et fermant après 19h, avec possibilité de prise en charge des devoirs |
| **Centres de loisirs** | Offre renforcée pendant les vacances scolaires (les jours fériés français ne coïncident pas avec ceux du Luxembourg ou de l'Allemagne) |
| **Services à domicile** | Développement des réseaux d'assistantes maternelles et de gardes partagées dans les zones frontalières |

**Recommandation** : Les collectivités territoriales des départements frontaliers (Moselle, Haut-Rhin, Meurthe-et-Moselle, Bas-Rhin) pourraient conditionner une partie de leurs financements aux structures d'accueil à l'adoption d'horaires élargis, en ciblant prioritairement les communes à forte proportion de travailleurs frontaliers.

### 6.4 Promouvoir le télétravail transfrontalier

La crise sanitaire de 2020-2021 a révélé le potentiel du télétravail pour les travailleurs frontaliers. Ce mode d'organisation, en réduisant le nombre de trajets hebdomadaires, pourrait lever une partie des contraintes identifiées dans cette étude — notamment la contrainte parentale et la pénibilité des trajets longs.

**Constat** : Le télétravail est particulièrement bénéfique pour les familles frontalières : un jour de télétravail par semaine représente une économie de 2h à 3h de trajet et une journée de présence au domicile facilitant la gestion des contraintes familiales. Pour les femmes, qui adoptent plus fréquemment le temps partiel comme stratégie de conciliation, le télétravail pourrait constituer une alternative permettant de maintenir un temps plein.

**Implication** : Le développement du télétravail transfrontalier suppose de lever plusieurs obstacles :

| Obstacle | Solution |
|----------|----------|
| **Fiscalité** | Clarification des conventions fiscales bilatérales sur le traitement du télétravail (seuils de jours, répartition de l'imposition) |
| **Protection sociale** | Harmonisation des règles de rattachement au régime de sécurité sociale (règlement européen 883/2004) |
| **Droit du travail** | Articulation entre le droit du travail du pays d'emploi et les conditions d'exercice au domicile (France) |
| **Équipement** | Accès au haut débit dans les zones rurales frontalières, espaces de coworking de proximité |

**Recommandation** : Les négociations en cours entre la France et ses voisins (notamment le Luxembourg, avec l'avenant fiscal de 2023 portant le seuil de télétravail à 34 jours) devraient être poursuivies et étendues, en visant un alignement des seuils entre les différents pays frontaliers.

### 6.5 Simplifier la complexité administrative pour les couples bi-frontaliers

Les couples bi-transfrontaliers — où les deux conjoints travaillent à l'étranger — font face à une complexité administrative accrue lorsque les deux membres travaillent dans des pays différents (par exemple, l'un au Luxembourg et l'autre en Allemagne).

**Constat** : Ces couples doivent naviguer entre plusieurs systèmes juridiques simultanément :
- **Fiscalité** : Deux pays d'emploi, une résidence fiscale en France, application de deux conventions bilatérales distinctes
- **Protection sociale** : Détermination du régime applicable (règlement européen), coordination des droits à la retraite, couverture maladie
- **Prestations familiales** : Règles de priorité entre pays, risque de doublons ou de lacunes

**Implication** : La simplification administrative bénéficierait particulièrement à ces couples. Plusieurs pistes :

| Domaine | Simplification recommandée |
|---------|----------------------------|
| **Déclaration fiscale** | Formulaires pré-remplis intégrant les données des employeurs étrangers, simulateurs en ligne |
| **Guichet unique** | Point de contact unique pour les questions fiscales, sociales et administratives des couples bi-frontaliers |
| **Coordination européenne** | Renforcement des échanges automatiques d'informations entre administrations (projet EESSI) |
| **Information** | Guides pratiques spécifiques aux couples bi-frontaliers, FAQ multilingues |

**Recommandation** : La création d'un statut ou d'un accompagnement spécifique pour les « couples bi-frontaliers » pourrait être envisagée au niveau des services déconcentrés de l'État (DIRECCTE, URSSAF, services fiscaux) et des organismes de coopération transfrontalière.

### 6.6 Intégrer la dimension genre dans les politiques d'accompagnement

L'hétérogénéité des effets selon le genre — 47 % des interactions significatives, temps partiel comme stratégie féminine, plafond de verre pour les indépendantes et diplômées — appelle une prise en compte explicite de la dimension genre dans les politiques d'accompagnement.

**Constat** : Les femmes ne sont pas dans la même position que les hommes face au travail frontalier. Elles font face à des contraintes spécifiques liées à la division sexuée du travail domestique, qui les conduisent soit à renoncer au travail frontalier, soit à adopter des stratégies de conciliation (temps partiel) moins rémunératrices.

**Implication** : Les politiques d'accompagnement devraient :

| Dimension | Action recommandée |
|-----------|-------------------|
| **Information** | Communication ciblée vers les femmes sur les opportunités frontalières, déconstruction des stéréotypes |
| **Formation** | Programmes de formation linguistique et professionnelle adaptés aux contraintes horaires des femmes (horaires décalés, modules courts, e-learning) |
| **Entrepreneuriat** | Accompagnement spécifique des femmes entrepreneures souhaitant développer une activité transfrontalière |
| **Temps partiel** | Information sur les droits au temps partiel dans les différents pays, simulation des impacts sur la retraite |

**Recommandation** : Les études d'impact des politiques transfrontalières devraient systématiquement intégrer une analyse différenciée selon le genre, conformément aux principes du *gender mainstreaming* promus par l'Union européenne.

### 6.7 Synthèse des recommandations

| # | Domaine | Recommandation | Niveau de décision |
|:-:|---------|----------------|:------------------:|
| 1 | Accompagnement | Approche « ménage » dans les services d'orientation | Régional / Local |
| 2 | Transport | Dimensionnement pour flux de couples, covoiturage organisé | Régional / Transfrontalier |
| 3 | Famille | Horaires élargis des structures d'accueil (6h-20h) | Local / Départemental |
| 4 | Télétravail | Extension des seuils fiscaux, harmonisation européenne | National / Bilatéral |

Ces recommandations, issues des résultats empiriques de l'étude, ne constituent pas des préconisations définitives mais des pistes de réflexion pour les acteurs publics. Leur mise en œuvre supposerait des analyses coûts-bénéfices approfondies et une concertation avec les parties prenantes (travailleurs frontaliers, employeurs, collectivités, partenaires transfrontaliers).

## 7. Forces de l'étude

### 7.1 Une base de données de grande ampleur

La première force de cette étude réside dans la **taille exceptionnelle de l'échantillon**. Avec 494 483 actifs occupés pour l'analyse individuelle et 126 935 couples bi-actifs pour l'analyse conjugale, cette recherche s'appuie sur l'une des plus grandes bases de données jamais constituées pour l'étude du travail transfrontalier en France.

| Échantillon | Effectif | Comparaison |

Cette ampleur confère une **puissance statistique élevée** permettant de détecter des effets de faible amplitude et d'estimer avec précision les coefficients. Les intervalles de confiance étroits (par exemple, ρ = 0,64 ± 0,02) témoignent de cette précision. Elle autorise également des analyses fines par sous-groupes (département, secteur, CSP) sans perte excessive de puissance.

## 8. Limites de l'étude

Toute recherche empirique comporte des limites qu'il convient d'expliciter pour permettre une interprétation éclairée des résultats. Cette section présente les principales limites de l'étude, organisées en cinq catégories : limites liées à l'identification causale, aux données, à la spécification des modèles, à la couverture géographique et à la généralisabilité.

### 8.1 Corrélation et non causalité

La limite la plus fondamentale de cette étude est l'**impossibilité d'établir des relations causales**. Les modèles estimés documentent des associations statistiques — des corrélations partielles entre variables explicatives et probabilité de travail transfrontalier — mais ne permettent pas d'affirmer que ces associations reflètent des effets causaux.

**Le problème de l'endogénéité** : Plusieurs variables explicatives sont potentiellement endogènes, c'est-à-dire que la causalité pourrait être inverse ou que des facteurs non observés pourraient causer simultanément la variable explicative et la variable dépendante.

| Variable | Endogénéité suspectée | Mécanisme |
|----------|----------------------|-----------|
| Motorisation | Forte | Les revenus frontaliers permettent l'achat de véhicules |
| Statut d'occupation | Forte | Les revenus frontaliers permettent l'accession à la propriété |
| Temps partiel | Moyenne | Choix simultané avec le travail frontalier |
| Secteur d'activité | Moyenne | Choix de secteur influencé par les opportunités frontalières |
| Localisation résidentielle | Forte | Installation près de la frontière *pour* travailler à l'étranger |

**Le cas du paramètre ρ** : Le coefficient de corrélation ρ = 0,64 du probit bivarié mesure une corrélation résiduelle entre les décisions des conjoints, *après contrôle* des caractéristiques observables. Cependant, cette corrélation ne permet pas de distinguer :

1. Un **effet causal direct** : le statut du conjoint *cause* la décision de l'autre (effet de réseau, transmission d'information)
2. Une **sélection sur inobservables** : les conjoints partagent des préférences non observées (tolérance aux trajets, appétence salariale) qui causent les deux décisions sans lien causal entre elles
3. Une **localisation endogène** : le couple a choisi de s'installer près de la frontière *parce que* les deux membres souhaitaient travailler à l'étranger

**L'absence d'instrument** : L'identification causale nécessiterait une source de variation exogène (instrument, discontinuité, expérience naturelle) affectant la probabilité de travail frontalier sans être corrélée aux caractéristiques non observées. Aucune telle source n'a été identifiée dans les données du Recensement. Les résultats doivent donc être interprétés comme des **associations conditionnelles**, non comme des effets causaux.

### 8.2 Données en coupe transversale

Le Recensement de la Population 2022 fournit une **photographie à un instant donné** de la situation des individus. Cette nature transversale impose plusieurs limites.

**Stock vs flux** : Le modèle estime la probabilité d'*être* transfrontalier à la date du Recensement, non celle de le *devenir*. Les résultats mélangent donc :
- Les déterminants de l'**entrée** dans le travail frontalier (qui devient transfrontalier ?)
- Les déterminants du **maintien** dans ce statut (qui reste transfrontalier ?)

Ces deux processus peuvent obéir à des logiques différentes. Par exemple, le diplôme pourrait faciliter l'entrée (accès aux emplois qualifiés) mais être moins déterminant pour le maintien (effet d'ancienneté).

**Absence de dynamique** : L'analyse ne permet pas d'étudier :
- L'**ordre des transitions** au sein du couple (qui devient transfrontalier en premier ?)
- Les **trajectoires** professionnelles (passage du travail local au travail frontalier, retours)
- Les **effets de durée** (le comportement diffère-t-il selon l'ancienneté dans le statut ?)

**Biais de survie** : L'échantillon ne contient que les individus qui sont transfrontaliers *et* qui résident toujours dans le Grand Est. Ceux qui ont quitté la région (déménagement vers le pays de travail, retour vers une autre région française) ne sont pas observés, ce qui peut biaiser les estimations.

### 8.3 Variables omises

Malgré la richesse des données du Recensement (65 variables exploitées), plusieurs déterminants potentiellement importants du travail transfrontalier ne sont pas observés.

| Variable omise | Importance supposée | Conséquence |
|----------------|---------------------|-------------|
| **Compétences linguistiques** | Élevée | L'allemand, le luxembourgeois, l'anglais sont essentiels pour travailler à l'étranger |
| **Revenus individuels** | Élevée | Le différentiel de salaire est le moteur économique du travail frontalier |
| **Distance précise à la frontière** | Élevée | Le département est une approximation grossière |
| **Réseaux professionnels** | Moyenne | Connaissance d'opportunités via des contacts |
| **Préférences non observées** | Élevée | Aversion au risque, tolérance aux trajets, ouverture internationale |
| **Historique professionnel** | Moyenne | Expérience antérieure à l'étranger |
| **Santé** | Faible à moyenne | Capacité à supporter des trajets longs |

**Biais de variable omise** : L'omission de ces variables peut biaiser les coefficients estimés si elles sont corrélées à la fois à la variable dépendante et aux variables explicatives incluses. Par exemple, si les individus de nationalité étrangère sont plus susceptibles de parler allemand (variable omise), l'effet estimé de la nationalité capte en partie l'effet des compétences linguistiques.

**Contribution à ρ** : Les variables omises communes aux deux conjoints (préférences partagées, réseaux familiaux, localisation fine) contribuent à la corrélation résiduelle ρ. Une partie de ρ = 0,64 reflète donc de l'hétérogénéité non observée plutôt qu'une coordination causale.

### 8.4 Limites de spécification

Plusieurs choix de spécification, bien que justifiés, imposent des limites à l'analyse.

**Hypothèse de linéarité dans l'index** : Le modèle probit suppose que les variables explicatives affectent la variable latente de manière linéaire (additive). Des non-linéarités ou des interactions complexes pourraient être mal captées. L'introduction d'interactions genre (Partie 4) répond partiellement à cette limite, mais d'autres interactions (département × secteur, âge × diplôme) n'ont pas été explorées.

**Homogénéité de ρ** : Le probit bivarié suppose un coefficient de corrélation ρ unique pour l'ensemble des couples. Or, la coordination conjugale pourrait varier selon :
- La présence d'enfants (contraintes de garde partagées)
- Le département (opportunités différentes selon les bassins d'emploi)
- Le pays de destination (Luxembourg vs Allemagne vs Suisse)
- L'ancienneté dans le couple

Un modèle avec ρ variable (probit bivarié à effets aléatoires) permettrait d'explorer cette hétérogénéité mais n'a pas été estimé.

**Sélection des couples bi-actifs** : L'échantillon du probit bivarié ne contient que les couples où **les deux membres sont actifs occupés**. Sont exclus :
- Les couples avec un conjoint inactif (retraité, au foyer, étudiant)
- Les couples avec un conjoint au chômage
- Les couples avec un conjoint en emploi partiel non déclaré

Cette sélection peut affecter la généralisabilité des résultats si les couples bi-actifs diffèrent systématiquement des autres couples.

### 8.5 Couverture géographique incomplète

**Exclusion des Ardennes** : Le département des Ardennes (08) a dû être exclu de l'analyse en raison d'une anomalie de codage dans le fichier MOBPRO (codes 80-85 au lieu de 08). Cette exclusion réduit la couverture des zones frontalières avec la Belgique.

| Impact | Évaluation |
|--------|------------|
| % transfrontaliers région concernés | ~1 % |
| Biais potentiel | Faible (flux vers Belgique minoritaires) |
| Généralisation vers Belgique | Limitée |

Les 9 départements retenus concentrent néanmoins plus de **98 % des travailleurs transfrontaliers** de la région Grand Est. L'exclusion des Ardennes ne compromet pas la validité des résultats pour les flux vers le Luxembourg, l'Allemagne et la Suisse, qui représentent la quasi-totalité du phénomène.

**Granularité départementale** : L'analyse géographique repose sur des indicatrices départementales, qui constituent une approximation grossière de la distance à la frontière. Au sein d'un même département, la situation peut varier considérablement :
- En Moselle, Thionville (à 15 km du Luxembourg) et Sarrebourg (à 80 km) n'offrent pas les mêmes opportunités
- En Haut-Rhin, Saint-Louis (frontière suisse) et Colmar (50 km) sont dans des situations différentes

Une analyse à l'échelle communale ou cantonale aurait été plus précise mais se serait heurtée à des effectifs insuffisants dans de nombreuses unités géographiques.

### 8.6 Généralisabilité

**Spécificité du Grand Est** : Les résultats sont obtenus sur les données du Grand Est et ne sont pas directement généralisables à d'autres régions frontalières françaises (Hauts-de-France, Auvergne-Rhône-Alpes, Nouvelle-Aquitaine). Les caractéristiques des marchés du travail frontaliers (Luxembourg vs Espagne vs Italie), les différentiels de salaires et les infrastructures de transport diffèrent significativement.

**Spécificité temporelle** : Le Recensement 2022 capture une situation post-Covid, après une période de développement massif du télétravail et de perturbation des flux frontaliers. Les comportements observés pourraient différer de ceux d'avant-crise ou d'une période « normale ».

**Spécificité des couples hétérosexuels** : L'analyse de la coordination conjugale se limite aux couples hétérosexuels (par construction, pour permettre l'identification REF/CONJ par le genre). Les résultats ne sont pas généralisables aux couples de même sexe, dont les dynamiques de négociation conjugale pourraient différer.

| # | Limite | Gravité | Atténuation |
|:-:|--------|:-------:|-------------|
| 1 | Corrélation ≠ causalité | **Élevée** | Interprétation prudente, pas de recommandations causales |
| 2 | Données transversales (stock) | Moyenne | Compléter par études longitudinales |
| 3 | Variables omises (langues, revenus) | Moyenne | Contribuent à ρ, biais possible sur certains coefficients |

Les associations statistiques identifiées sont robustes et informatives pour la compréhension du phénomène ; elles ne constituent pas, en revanche, des preuves de relations causales exploitables pour des interventions de politique publique sans précaution.

## 9. Pistes de prolongement

Les résultats de cette étude ouvrent plusieurs pistes de recherche qui permettraient d'approfondir la compréhension des déterminants du travail transfrontalier et de dépasser certaines des limites identifiées.

### 9.1 Exploitation de données longitudinales

Le passage à des **données de panel** constituerait l'extension la plus naturelle de ce travail. Plusieurs sources pourraient être mobilisées :

| Source | Apport potentiel | Faisabilité |
|--------|------------------|:-----------:|
| **Échantillon Démographique Permanent (EDP)** | Suivi individuel sur plusieurs recensements, transitions vers/depuis le travail frontalier | ★★★ |
| **Données fiscales appariées (FILOSOFI)** | Revenus individuels, différentiels de salaires, dynamique des revenus | ★★☆ |
| **DADS/DSN appariées** | Historique employeur, ancienneté, trajectoires professionnelles | ★★☆ |
| **Enquête Emploi en continu** | Panel rotatif, transitions trimestrielles | ★☆☆ (effectifs frontaliers faibles) |

L'exploitation de données longitudinales permettrait de répondre à des questions inaccessibles en coupe transversale :
- **Qui devient transfrontalier ?** Distinction entre déterminants de l'entrée et du maintien
- **Dans quel ordre au sein du couple ?** Séquence temporelle des décisions conjugales — qui est le « pionnier », qui est le « suiveur » ?
- **Quels sont les facteurs déclencheurs ?** Événements de vie (naissance, divorce, licenciement, achat immobilier) précédant la transition
- **Y a-t-il des effets d'hystérèse ?** Une fois transfrontalier, le retour vers l'emploi domestique est-il fréquent ?

### 9.2 Hétérogénéité du paramètre ρ

Le modèle bivarié actuel suppose un paramètre de corrélation **unique** (ρ = 0,64) pour l'ensemble des couples. Cette hypothèse d'homogénéité pourrait être relâchée en estimant des modèles à **ρ variable** selon les caractéristiques du couple.

| Dimension d'hétérogénéité | Hypothèse testable |
|---------------------------|-------------------|
| **Présence d'enfants** | ρ plus faible pour les couples avec enfants (contrainte de spécialisation) vs sans enfants (coordination facilitée) |
| **Département de résidence** | ρ plus élevé en Moselle (proximité Luxembourg, concentration des opportunités) qu'en Bas-Rhin (marché allemand plus diffus) |
| **Pays de destination** | ρ différent selon que les conjoints travaillent dans le même pays ou dans des pays différents |
| **Niveau de diplôme** | ρ plus élevé pour les couples de cadres (réseaux professionnels partagés) que pour les couples d'ouvriers |
| **Ancienneté dans le statut** | ρ croissant avec l'ancienneté (renforcement de la coordination au fil du temps) |

Techniquement, cette extension supposerait l'estimation de modèles probit bivariés à coefficients aléatoires ou l'introduction d'interactions dans la spécification de ρ. Ces approches, plus complexes, permettraient de mieux comprendre **pour qui** la coordination conjugale est la plus forte.

### 9.3 Identification causale

L'établissement de relations causales supposerait l'exploitation de **variations exogènes** affectant le travail transfrontalier indépendamment des caractéristiques individuelles. Plusieurs stratégies pourraient être envisagées :

**Variables instrumentales** :
- Distance euclidienne à la frontière (instrument géographique)
- Présence historique d'industries transfrontalières dans la commune (instrument historique)
- Densité de travailleurs frontaliers dans le voisinage (instrument de pairs)

**Expériences naturelles** :
- Ouverture d'une nouvelle ligne de transport transfrontalière (TGV Est, tram Strasbourg-Kehl)
- Modification d'une convention fiscale bilatérale (avenant Luxembourg 2023 sur le télétravail)
- Fermeture d'un grand employeur frontalier (choc négatif)
- Crise sanitaire COVID-19 et fermeture temporaire des frontières (choc symétrique)

**Régression sur discontinuité** :
- Discontinuité spatiale à la frontière départementale (Moselle vs Meurthe-et-Moselle)
- Discontinuité temporelle lors de changements réglementaires

Ces stratégies permettraient de passer d'une logique descriptive (« les frontaliers ont telles caractéristiques ») à une logique causale (« devenir frontalier *cause* tel effet »).

### 9.4 Modélisation du choix du pays de destination

Ce travail a volontairement exclu la question du **pays de destination**, considérée comme fortement endogène à la localisation résidentielle. Une extension naturelle consisterait à modéliser ce choix via un **modèle multinomial emboîté** (nested logit) :

```
                    Travail
                      │
            ┌─────────┴─────────┐
            │                   │
        En France          À l'étranger
                                │
                    ┌───────┬───┴───┬───────┐
                    │       │       │       │
                Luxembourg  Allemagne  Suisse  Belgique
```

**Niveau 1** : Choix binaire travail en France vs à l'étranger (probit binaire actuel)
**Niveau 2** : Choix du pays parmi les destinations possibles (logit multinomial conditionnel)

Cette structure emboîtée permettrait de tester si les déterminants de la décision « transfrontalier ou non » diffèrent de ceux du choix du pays. Par exemple :
- Le diplôme favorise-t-il davantage le Luxembourg (secteur financier) que l'Allemagne (industrie) ?
- La nationalité allemande oriente-t-elle vers l'Allemagne plutôt que vers le Luxembourg ?
- Les femmes sont-elles surreprésentées vers certains pays (Suisse, secteur de la santé) ?

### 9.5 Capturer l'attractivité fiscale et économique des pays frontaliers

Une limite importante de l'analyse actuelle est l'absence de variables mesurant directement l'**attractivité économique et fiscale** des pays de destination. Les effets départementaux captent indirectement cette attractivité (la Moselle est proche du Luxembourg, le Haut-Rhin de la Suisse), mais ne permettent pas de distinguer ce qui relève de la distance géographique de ce qui relève des incitations économiques.

Plusieurs extensions permettraient de capturer explicitement ces effets :

**Variables d'attractivité économique** :

| Variable | Source potentielle | Hypothèse |
|----------|-------------------|-----------|
| **Différentiel de salaire médian** | Eurostat, offices statistiques nationaux | Plus le différentiel est élevé, plus la probabilité de travail frontalier augmente |
| **Taux de chômage du pays voisin** | Eurostat | Un faible taux de chômage frontalier signale des opportunités d'emploi |
| **Croissance du PIB frontalier** | Comptes nationaux | Dynamisme économique = demande de travail |
| **Spécialisation sectorielle** | Comptes nationaux, registres d'entreprises | Adéquation entre profil individuel et structure productive du pays |

**Variables d'attractivité fiscale** :

| Variable | Source potentielle | Hypothèse |
|----------|-------------------|-----------|
| **Taux marginal d'imposition** | Législations fiscales nationales | Fiscalité avantageuse → incitation au travail frontalier |
| **Conventions fiscales bilatérales** | Textes juridiques | Régime d'imposition (à la source vs résidence) affecte le gain net |
| **Cotisations sociales** | Législations sociales | Coût du travail et prestations associées |
| **Seuils de télétravail autorisés** | Avenants fiscaux récents | Flexibilité du travail frontalier |

**Stratégie empirique** : L'introduction de ces variables supposerait de travailler à un niveau géographique fin (commune ou canton) permettant de calculer la distance à chaque frontière et d'affecter à chaque individu l'attractivité du pays le plus proche. Un modèle à choix discret spatial pourrait alors estimer l'élasticité de la probabilité de travail frontalier au différentiel de salaire ou au taux d'imposition.

**Intérêt pour les politiques publiques** : Cette extension permettrait de simuler l'impact de changements fiscaux ou économiques sur les flux de travailleurs frontaliers. Par exemple :
- Quel serait l'effet d'une harmonisation fiscale européenne sur les flux vers le Luxembourg ?
- Comment une récession en Allemagne affecterait-elle les frontaliers alsaciens ?
- Le relèvement du seuil de télétravail (34 jours vers le Luxembourg) a-t-il modifié les comportements ?

### 9.6 Analyse spatiale fine

L'intégration de variables géographiques plus précises affinerait la compréhension du gradient spatial :

| Variable | Apport |
|----------|--------|
| **Distance euclidienne à la frontière** | Mesure continue vs indicatrices départementales discrètes |
| **Temps d'accès au premier bassin d'emploi étranger** | Intègre la qualité des infrastructures de transport |
| **Accessibilité routière (isochrones)** | Zones atteignables en 30, 45, 60 minutes |
| **Densité de frontaliers dans la commune** | Effets de pairs, réseaux locaux |
| **Historique industriel transfrontalier** | Path dependence, culture frontalière locale |

Ces variables permettraient de dépasser la maille départementale et d'identifier des **effets de seuil** (au-delà de quelle distance le travail frontalier devient-il improbable ?) ou des **effets de concentration** (existe-t-il des clusters de communes à forte proportion de frontaliers ?).

### 9.7 Extension à d'autres régions frontalières

Le cadre méthodologique développé ici pourrait être **répliqué** sur d'autres régions frontalières françaises :

| Région | Pays frontaliers | Spécificités |
|--------|------------------|--------------|
| **Hauts-de-France** | Belgique | Proximité culturelle et linguistique, intégration européenne |
| **Auvergne-Rhône-Alpes** | Suisse, Italie | Genève (finance), travail saisonnier (tourisme) |
| **Provence-Alpes-Côte d'Azur** | Italie, Monaco | Monaco = cas particulier (fiscalité, proximité) |
| **Occitanie** | Espagne, Andorre | Flux plus faibles, intégration transfrontalière moindre |
| **Nouvelle-Aquitaine** | Espagne | Pays basque, dynamique économique récente |

La comparaison inter-régionale permettrait de tester la **généralisabilité** des résultats (le paramètre ρ est-il similaire partout ?) et d'identifier des **spécificités locales** (le plafond de verre est-il plus marqué vers certains pays ?).

### 9.8 Intégration de données qualitatives

Les méthodes quantitatives mobilisées dans cette étude pourraient être enrichies par des **approches qualitatives** complémentaires :

| Méthode | Apport |
|---------|--------|
| **Entretiens semi-directifs** | Comprendre les motivations subjectives, les arbitrages familiaux, les représentations du travail frontalier |
| **Récits de vie** | Retracer les trajectoires individuelles et conjugales, identifier les moments-clés |
| **Focus groups** | Explorer les perceptions collectives, les normes sociales, les effets de pairs |
| **Ethnographie** | Observer les pratiques quotidiennes (trajets, organisation familiale, réseaux) |

Ces approches permettraient d'éclairer les **mécanismes** sous-jacents aux régularités statistiques. Pourquoi les femmes adoptent-elles le temps partiel comme stratégie de conciliation ? Comment les couples coordonnent-ils concrètement leurs décisions ? Quels sont les freins psychologiques ou culturels au travail frontalier ?

### 9.9 Tableau synthétique des prolongements

| # | Prolongement | Question de recherche | Données/Méthode |
|:-:|--------------|----------------------|-----------------|
| 1 | Données longitudinales | Qui *devient* transfrontalier ? Séquence des décisions conjugales ? | EDP, DADS, panel |
| 2 | Hétérogénéité de ρ | Pour qui la coordination est-elle la plus forte ? | Probit bivarié à ρ variable |
| 3 | Identification causale | Quel est l'*effet causal* du travail frontalier ? | IV, expériences naturelles, RDD |

Ces prolongements, ambitieux, dépassent le cadre d'un mémoire de Master mais dessinent un **programme de recherche** sur le travail transfrontalier en France — phénomène massif (près de 450 000 frontaliers au niveau national) mais encore insuffisamment étudié par la recherche académique.

## 10. Mot de la fin

Au terme de ce travail, une conviction s'impose : le travail transfrontalier dans le Grand Est ne peut être compris comme la simple agrégation de décisions individuelles.

La géographie dessine le champ des possibles — résider en Moselle plutôt qu'en Marne augmente de 31 points la probabilité d'être frontalier, effet sans commune mesure avec tous les autres déterminants. Les réseaux transnationaux ouvrent les portes — nationalité étrangère, naissance hors de France, familiarité avec la mobilité internationale. Le profil socioprofessionnel filtre les candidats — le secteur public et l'agriculture sont structurellement exclus, les ouvriers qualifiés et les cadres surreprésentés. Les contraintes familiales pèsent différemment selon le genre — le temps partiel comme stratégie féminine de conciliation, la pénalité parentale qui s'intensifie pour les femmes au-delà de deux enfants.

Mais c'est le résultat sur la coordination conjugale qui constitue, à nos yeux, l'apport le plus significatif de cette étude. Le paramètre ρ = 0,64 — traduit par un effet multiplicateur de près de cinq — établit que les choix professionnels des conjoints ne sont pas indépendants mais s'inscrivent dans une **logique de ménage**. Avoir un conjoint transfrontalier fait passer la probabilité de l'être soi-même de 6,6 % à 47,5 %. Les couples bi-transfrontaliers, 12,8 fois plus fréquents qu'attendu sous hypothèse d'indépendance, ne sont pas une anomalie statistique mais le résultat d'une stratégie familiale coordonnée.

Cette dimension conjugale, jusqu'alors absente des analyses économétriques sur données françaises, invite à repenser les politiques d'accompagnement du travail frontalier. Dans une région où plus de 200 000 actifs traversent quotidiennement une frontière, comprendre que ces décisions se prennent à deux — et non seul — n'est pas un détail méthodologique. C'est une clé de lecture indispensable pour appréhender les dynamiques territoriales du Grand Est et accompagner efficacement les ménages frontaliers dans leurs choix professionnels et familiaux.
